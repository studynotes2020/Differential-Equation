# Solve

$$y' + ky = kq(t)$$

with $k > 0$

Solution: $$y = e^{-kt}\int q(t)e^{kt}\ dt + ce^{-kt}$$

$ce^{-kt} \to 0$ as $t \to \infty$ so this is called transient solution. $c$ is calculated from the initial condition

$e^{-kt}\int q(t)e^{kt}\ dt$ this term stays, so steady state solution or long term solution

Hence for these type of equations, they can be started anywhere but will approach the steady state solution as time goes by as transient solution approaches zero

Input (or right hand side): $q(t)$

Response (solution to ODE): $y(t)$

So, above equation can also be written as 

$$\frac{1}{k}y' + y = q(t)$$


## Superposition of Inputs

If $q_1(t)$ produces response $y_1(t)$ and $q_2(t)$ produces response $y_2(t)$

then $q_1(t) + q_2(t)$ produces response $y_1(t) + y_2(t)$ and

$cq_1(t)$ produces response $cy_1(t)$

This uses the fact that the equation is linear 




## Back to the ODE

$$y' + ky = kq(t)$$ and $q(t) = \cos \omega t$

$\omega = $ angular frequency = the number of complete oscillation in $2\pi$

Problem: given input $q(t) = \cos\omega t$, find the response

Strategy: Complexify the problem, solve that, then go back to the real domain. This is because it is easier to integrate exponential

View the $\cos wt$ as real part of $e^{iw t}$ and let $\tilde y = y_1 + iy_2$ which is the complex solution. Then the equation becomes

$$\tilde y' + k\tilde y = ke^{iwt}$$

Find $\tilde y$ and the solution of the original problem is the real part of the solution, in this case, $y_1$

Multiply both side with integrating factor, $e^{kt}$ and simplify

$$(\tilde y e^{kt})' = ke^{(k + iw)t}$$

$$\tilde ye^{kt} = \frac{k}{k + iw}e^{(k + iw)t}$$

$$\tilde y = \frac{1}{1 + i(w/k)}e^{iwt}$$

Take the real part, two methods
1. Go polar form
2. Go cartesian

Remind for $a$ is complex: 

$\frac{1}{a}.a = 1$ 

$|\frac{1}{a}| = \frac{1}{|a|}$

$arg(\frac{1}{a}) + arg(a) = 0$, $arg(1) = 0$

$arg(\frac1a) = -arg(a)$

Polar form: $$\frac{1}{1 + i(w/k)} = Ae^{-i\phi}$$

To convert cartesian to polar form above, think right triangle with sides 1 and $\frac{w}{k}$. Then $A$ is the hypotenuse and $\phi = arg(1 + i(w/k))$, so its reciprocal is $-\phi$

$$Ae^{-i\phi} = \frac{1}{\sqrt{1 + (w/k)^2}}e^{-i\phi}$$

Substitute back

$$\tilde y = Ae^{-i\phi} e^{iwt} = \frac{1}{\sqrt{1 + (w/k)^2}}e^{i(wt -\phi)}$$

The response is the real part of the answer $$y_1 = Re(\tilde y) = \frac{1}{\sqrt{1 + (w/k)^2}} \cos(wt - \phi)$$

$\phi = \tan^{-1} (w/k)$ phase lag of the function

From the solution, conductivity k goes up, denominator smaller, amplitude A goes up.

Also, from the triangle, if the side w/k goes down, $\phi$ goes down

To explain physically, when conductivity goes up, heat goes in much faster, so amplitude rises quickly and very little lag to response to the input
