# Substitution Method

So far two general methods to solve First Order Differential Equation
1. Separation of variables
2. Integrating factor

In some sense, there are only two general methods for those kinds of equations that can always be solved.

For other equations, the solution is done by changing the variables of the equation to the case we already know what to solve

## Scaling

$$y' = f(x, y)$$

$$ x_1 = \frac{x}{a}, y_1 = \frac{y}{b}$$

$a, b$ is constants

Reason of Scaling
1. Change units
2. Make variable dimensionless (without units)
3. Reduce the number, or simplify constants

$$\frac{dT}{dt} = k(M^4 - T^4)$$

This equation is valid for big temperature difference. T is the internal temperature, M is the constant external temperature

Scale by relating T to M: $$T_1 = \frac{T}{M}$$

$$MT_1 = T$$

$$M\frac{dT_1}{dt} = kM^4(1-T_1^4)$$

$$\frac{dT_1}{dt} = k_1(1-T_1^4)$$

The equation is one less constant, where $k_1 = kM^3$ has unit inverse time. It is called "lumping constant". $(time)^{-1}$ and $T_1$ is dimensionless.




## Direct Substitution

new variable = some combinations of old variables

Example: $T_1 = T/M$

When doing integration $\int x\sqrt{1 - x^2}\ dx$, substitute $u = 1 - x^2$

## Inverse Substitution

old variable = some combination of new and old

Example: $T = MT_1$

$\int \sqrt{1 - x^2}\ dx$, substitue $x = sin\ u$


## Direct Substitution to Solve this kind

$$y' = p(x)y + q(x)y^n$$

$n$ can be anything but zero and one. If n is 1, just use separation variable

Divided by $y^n$ becomes Bernoulli equation

$$\frac{y'}{y^n} = p(x)\frac{1}{y^{n-1}} + q(x)$$

Let $v = \frac{1}{y^{n-1}} = y^{1 - n}$, then $v' = (1 - n)y^{-n}y'$

$$\frac{v'}{1- n} = p(x)v + q(x)$$

Solve this like standard form first order to get v, then get y

### Example

$$y' = \frac{y}{x} - y^2$$

$$\frac{y'}{y^2} = \frac{1}{x}\frac{1}{y} - 1$$

Substitute $v = \frac{1}{y}, v' = -\frac{1}{y^2}y'$

$$-v' = \frac{v}{x} - 1$$

$$v' + \frac{v}{x} = 1$$

Integrating factor: $e^{ln\ x} = x$

$$xv' + v = x$$

$$(xv)' = x$$

$$xv = \frac{1}{2}x^2 + C$$

$$v = \frac{1}{2}x + \frac{C}{x} = \frac{x^2 + 2C}{2x}$$

$$y = \frac{1}{v} = \frac{2x}{x^2 + 2C}$$

## Inverse Substitution to Solve Homogeneous ODE

Homogeneous ODE: $$y' = F(y/x)$$

Check if the equation is homogeneous

$$y' = \frac{x^2y}{x^3 + y^3} = \frac{y/x}{1 + (y/x)^3}$$

so this equation is homogeneous. How about this one?

$$xy' = \sqrt{x^2 + y^2}$$

$$y' = \sqrt{1 + (y/x)^2}$$

This is homogeneous

### Invariant under zoom

zoom means $x = ax_1$, $y = ay_1$

$dx = a\ dx_1$, $y = a\ dy_1$

$$\frac{dy_1}{dx_1} = F(\frac{y_1}{x_1})$$

Every y and x change to $y_1, x_1$, so the equation doesn't changed via zoom transformation

## How to solve homogeneous ODE

$$y' = F(y/x)$$

substitute $z = y/x \implies y = zx$, $y' = z'x + z$

$$z'x + z = F(z)$$

$$x\frac{dz}{dx} = F(z) - z$$

Then proceed to separation of variables and solve it

### Example

Problem: The boat want to go at constant 45 degree from the helicopter beam while the beam is trying to aim at the boat

What's the boats' path?
We want slope (moving path) makes constant known angle

Let $y = y(x)$ is the unknown function which is the boat path. $y' = tan\ (\theta + 45^o)$

And slope of the beam is $tan\ \theta = y/x$

$$y' = tan\ (\theta + 45^o) = \frac{tan\ \theta + tan\ 45^0}{1 - tan\ \theta \ tan\ 45^0}$$

$$y' = \frac{y/x + 1}{1 - y/x} \text{homogeneous ODE}$$

$z = y/x$, $y' = z'x + z$

$$z'x + z = \frac{z + 1}{1 - z}$$

$$x\frac{dz}{dx} = \frac{z + 1}{1 - z} - z = \frac{1 + z^2}{1 - z}$$

$$\frac{1 - z}{1 + z^2}dz = \frac{dx}{x}$$

$$tan^{-1}z - \frac12 ln\ (1 + z^2) = ln\ x + C$$

$$tan^{-1}z = ln\ \sqrt{1 + z^2} + ln\ x + C$$

$$tan^{-1}(y/x) = ln\ \sqrt{x^2 + y^2} + C$$

Change this to polar coordinate

$$\theta = ln\ r + C$$

Or $$r = C_1 e^{\theta}$$
