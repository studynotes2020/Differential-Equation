# The Basic Problem is to Solve

$$y' + ky = k\cos wt$$

$$\tilde y' + k\tilde y = ke^{iwt}$$

The solution is: $$\tilde y = \frac{1}{1 + i(w/k)}e^{iwt}$$

### Going polar

$$\tilde y = \frac{1}{\sqrt{1 + (w/k)^2}}e^{i(wt - \phi)}$$

Taking real part, $$y_1 = \frac{1}{\sqrt{1 + (w/k)^2}} \cos(wt - \phi)$$

### Going Cartesian

$$\tilde y = \frac{1}{1 + i(w/k)}e^{iwt} = \frac{1 - i(w/k)}{1 + (w/k)^2} (\cos wt + i\sin wt)$$ 

Taking real part $$y_1 = \frac{1}{1 + (w/k)^2} (\cos wt + \frac{w}{k} \sin wt)$$

From trigonometry identities
$$a\cos\theta + b\sin\theta = c \cos(\theta - \phi) $$

Think triangle with side a and b, hypertenuse c and $\phi = \tan^{-1}(b/a)$

$$y_1 = \frac{1}{1 + (w/k)^2}\sqrt{1 + (w/k)^2}\cos(wt - \phi)$$

We can see this solution is exactly same as polar form, except we need trigonometry identity for this method

## Proof of Formula

$$a\cos\theta + b\sin\theta = c \cos(\theta - \phi) $$

1. High school proof. Expand RHS with trigonometry identities, then we get LHS. The problem with this proof is that it is obvious RHS can be converted to LHS. But when we starts with LHS, it is unclear that we can convert to RHS

2. 18.02 proof. Realize that LHS is a dot product of vector $(a, b).(\cos\theta \sin\theta)$. From dot product of geometric definition
$$(a, b).(\cos\theta \sin\theta) = |(a,b)|.1.\cos(\theta - \phi) = c \cos(\theta - \phi)$$

3. 18.03 proof. Think LHS as real part of $(a - bi)(\cos\theta + i\sin\theta)$. 

$$a\cos\theta + b\sin\theta = Re[(a - bi)(\cos\theta + i\sin\theta)]$$

Expand this expressionon RHS 

$$(a - bi) = \sqrt{a^2 + b^2}e^{-i\phi}$$

$$\cos\theta + i\sin\theta = e^{i\theta}$$. 

Their product is $$\sqrt{a^2 + b^2}e^{i(\theta - \phi)}$$

and its real part is $$Re[(a - bi)(\cos\theta + i\sin\theta)] = Re[\sqrt{a^2 + b^2}e^{i(\theta - \phi)}] = \sqrt{a^2 + b^2} \cos(\theta - \phi) $$

## Basic Linear ODE

1. Temperature-Concentration, conduction-diffusion model

$$y' + ky = kq(t)$$

2. $y' + ky = q(t)$

3. Most general case, allow k to be non-constant

$$y' + p(t)y = q(t)$$

All three cases above $k > 0$


### Mixing Example (application of $y' + ky = kq(t)$)

Suppose there is a tank with salt flowing in, mixing and flowing out

$x(t) = $ amount of salt in tank at time t. I am interested to know how it varies with time. 

$C_e, v, r$ is the input data

$C_e =$ concentration of incoming salt, $v = $ volume of the tank, $r = $ flow rate

The problem is what happen to x of t

$\frac{dx}{dt} = $ rate of salt inflow - rate of salt outflow

$$\frac{dx}{dt} = rC_e - r\frac{x}{v}$$

Write the equation in standard form 

$$\frac{dx}{dt} + r\frac{x}{v} = rC_e$$

We can solve in this form, but to understand better what is happening, we express the term in concentration

$$\frac{dx}{dt} + rC = rC_e$$

The tank concentration $$C = \frac{x}{v}$$

Tank concentration is also a function of t

$$v\frac{dC}{dt} + rC = rC_e$$

In standard form

$$\frac{dC}{dt} + \frac{r}{v}C = \frac{r}{v}C_e$$

We can see it falls into first category, $y' + ky = kq(t)$. The analog of conductivity and diffusivity in here is $k = \frac{r}{v}$. r is rate of inflow, then r/v is the fractional rate of outflow. In other words if r/v is 1/10 per minute, then 1/10 of the tank will be emptied in one minute.

Unit of r is volume per min and unit of v in volume, hence unit of k is per min or reciprocal time which is the same units  of conductivity and diffusivity

Suppose $C_e$ is sinusoidal, e.g. $\cos wt$, how closely does the concetration of the tank $C(t)$ follow $C_e(t)$

If k is big, the inner concentration follow close the external concentration

If r is fast, liquid flow in fast, so the inner concentration follow closely the external concentration

If v i small, the tank got emptied quickly, hence the inner concentration follow closely the external concentration. Both are intuitive result

Closely followed means phase lag is small, amplitude A near 1

## Example application of $y' + ky = q(t)$

Suppose a circuit with power source E, a resistor R and a capacitor C

q = charge of the capacitor

$$\frac{dq}{dt} = i$$

$$R\frac{dq}{dt} + \frac{q}{C} = E(t)$$

$$q' + \frac{q}{RC} = \frac{E}{R}$$

People do not convert it to first category and keep in this form as in second category, and call k = 1/RC. 

Another example is radioactive decay

A decay to B and further decay

$$A \to B \to $$

$$\frac{dB}{dt} = k_1A - k_2B$$

To avoid two dependent constants

$$\frac{dB}{dt} = k_1A_0e^{-k_1t} - k_2B$$

$$B' + k_2B = k_1A_0e^{-kt}$$

This is the form in second category



### k is positive vs negative

If $k < 0$, none of the steady state, transient, input, response applies. The technique to solve the equation is the same

$$y' - ay = q(t)$$ where $a > 0$ so the coefficient of y is negative

The solution will be: $$e^{at}\int q(t)e^{-at}\ dt + Ce^{at}$$

Notice that transient solution goes to infinity or minus infinity. So it is not transient and heavily dependant on initial condition. And the steady state solution is a fixed function. Initial condition after a while doesn't matter anymore.

$k < 0$ in biology and economics or when life is involved. $k > 0$ when inanimate things are involved
