# Laplace Transform comes from power series

Power series

$$\sum_0^\infty a(n)x^n = A(x)$$

$a(n)$ gives the sequence of coefficients of the power series

$$a(n) \to A(x)$$

$$a(n) = 1 \to \frac{1}{1-x}, |x| < 1$$ 
All coefficient is 1 and that will produce a geometry series if |x| < 1. If |x| > 1, the series diverge.

$$a(n) = \frac{1}{n!} \to e^x$$

x doesn't need to be qualified because this is true for all of x. Taking a discrete function defined for non-negative integers, plug into the formula becomes a continuous function.

Variable n goes in, variable x comes out

# Continuous analog

Replace n = 0, 1, .. $\infty$ by t = 0, 1, ...  $\infty$

Similarly, replace summation by integration

$$\int_0^\infty a(t)x^t dt = A(x) $$

exponential is really the thing people want to differential and integrate, so convert x to $e^{ln x}$

$$x^t = (e^{ln x})^t$$

x needs to be smaller than 1. Also, if allow x negative, negative number power something will get into complex number. We don't want complex number,
so put $ 0 < x < 1$, which means

$$ \infty < ln x < 0$$

We would like to simplify more, hence make $-s = ln x$. Make it negative so that s always positive, more convinient to work with

Also, change $a(t)$ to $f(t)$ which is more conventional

After making all those changes

$$\int_0^\infty f(t)e^{-st} dt = A(e^{-s}) $$

Since we call input function $f(t)$, we can call output function $F(s)$

$$\int_0^\infty f(t)e^{-st} dt = F(s) $$

This is the continuous analogue of the summation of a power series. It starts with a function defined for $0 \leq t < \infty$ and turns it into a function of s

This is called Laplace transform


The variable get changed in the output function
$$f(t) \to tranform \to F(s)$$

The function comes out use the same variable. For example, differentiation
$$f(t) \to operator \to g(t)$$


## Linear Transform

Laplace transform is a linear transform

$$L(f + g) = L(f) + L(g)$$

$$L(cf) = cL(f)$$

## Laplace Transform for various function

$$L(0) = 0$$

$$\begin{align} L(1) & = \int_0^\infty e^{-st}\ dt \\
& = \frac{e^{-st}}{-s}|_0^\infty \\ 
& = \frac{0 - 1}{-s} \\
& = \frac{1}{s}, s > 0
\end{align}$$

because if $s < 0$, the integration diverges

$$\begin{align} L(e^{at}f(t)) & = \int_0^\infty e^{at}f(t)e^{-st}\ dt \\
& = f(t)e^{-(s - a)t}\ dt \\
& = F(s - a)
\end{align}$$

This is true when $s > a \implies s - a > 0$. So when you multiply exponential function on the left, you get shift of a on the right. This is called exponential-shift formula. When $f(t) = 1$, we get the result

$$L(e^{at}) = \frac{1}{s - a}$$

If a is complex number, this also works

$$L(e^{(a + bi)t} = \frac{1}{s - (a + bi)} $$

LP for sine and cosine. From Euler formula

$$\begin{align} L(\cos at)  & = L(\frac{e^{iat} + e^{-iat}}{2}) \\
& = \frac12 (\frac{1}{s - ia} + \frac{1}{s + ia}) \\
& = \frac12 \frac{2s}{s^2 + a^2} \\
& = \frac{s}{s^2 + a^2}, s > 0
\end{align}$$

Similarly for sine using Euler formula

$$L(\sin at) = \frac{a}{s^2 + a^2}. s > 0$$

Polynomial with integration by parts

$$\begin{align}L(t^n) & = \int_0^\infty t^n e^{-st}\ dt \\
& = t^n\frac{e^{-st}}{-s}|_0^\infty - \int_0^\infty nt^{n -1}\frac{e^{-st}}{-s}\ dt \\
\end{align}$$

What is this limit $$\lim \limits_{t \to \infty} \frac{t^n}{-se^{st}}$$

$e^{st}$ wins the race over $t^n$. L'Hopital rule repeatedly t will become zero, but denominator still has exponential, hence 

$$\lim \limits_{t \to \infty} \frac{t^n}{-se^{st}} = 0$$

$$L(t^n) = 0 - 0 + \frac{n}{s}\int_0^\infty t^{n - 1}e^{-st}\ dt = \frac{n}{s}L(t^{n - 1})$$

We get a reduction

$$L(t^n) = \frac{n}{s} L(t^{n - 1}) = \frac{n(n-1)..1}{s!}L(t^0) = \frac{n!}{s^{n+1}}$$

## Inverse Laplace Transform

What is the inverse LP for $\frac{1}{s(s + 3)}$

$$\begin{align} L^{-1}(\frac{1}{s(s + 3)}) & = L^{-1}(\frac{1/3}{s} + \frac{-1/3}{s + 3}) \\ 
& = \frac13 + \frac{1}{3}e^{-3t}
\end{align}$$
