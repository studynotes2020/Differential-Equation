## Temperature-concentration model

### Conduction

Suppose a vacuum inside a water tank and a box inside the vacuum. T for temperature inside the box, $T_e$ is the temperature of the water. Temperature of water goes into the temperature of the box only by conduction.

Newton law of cooling. $T$ is the temperature inside, $T_e$ is the temperature outside which is bigger and $t$ is the time. We want to write it so that $k > 0$ where k is the conductivity.

$$\frac{dT}{dt} = k(T_e - T)$$

Initial condition $T(0) = T_0$

### Diffusion

$C =$ salt concentraion inside. $C_e =$ salt concentration outide. The salt diffuse from outside into inside (membrane wall)

$$\frac{dC}{dt} = k(C_e - C)$$

We expect salt to flow in, so $k > 0$

These two differential equation models the physical situation of conduction and diffusion.



First converts it to standard linear form

$$T' + kT = kT_e$$

$T_e$ can be a general function of $T$. Here $k$ is constant, but this is not necessary, it can be varying over time.

Back to the standard form: $$y' + p(x)y = q(x)$$

We can solve it by integrating factor: $u(x)$. Multiply the equation with this factor

$$uy' + puy + qu$$

So that we can write the equation as follow

$$(uy)' = qu$$

Then we can solve it. In order to achive this, u' has to be equal to pu $$u' = pu$$

Integrate with separation variable

$$ \int \frac{u'}{u}\ du = \int p(x)\ dx$$
$$ln\ u = \int p(x)\ dx$$
$$u = e^{\int p\ dx}$$

This is called the integrating factor. (No arbitrary constant, since only one $u$ you needed)

## Method

1. Put it in standard linear form $y' + py = q$
2. Calculate integrating factor $e^{\int p\ dx}$
3. Multiply both sides by $e^{\int p\ dx}$
4. Integrate

## Example

$$xy' - y = x^3$$

Put it in standard form

$$y' - \frac{1}{x}y = x^2$$

$\int 1/x\ dx = -ln\ x$, hence integrating factor, $$e^{-ln\ x} = (e^{ln\ x})^{-1} = \frac1x$$

Multiply both sides by integrating factor

$$\frac{1}{x}y' - \frac{1}{x^2}y = x$$

$$(\frac1x y)' = x$$

Do the integration

$$\frac{y}{x} = \frac{x^2}{2} + C$$

Solution is

$$y = \frac{x^3}{2} + Cx$$

## Example

$$(1 + \cos x)y' - (\sin x)y = 2x$$

$$y' - \frac{\sin x}{1 + \cos x}y = \frac{2x}{1 + \cos x}$$

Look for integrating factor: $$-\int \frac{\sin x}{1 + \cos x}\ dx = ln\ (1 + \cos x)$$

Integrating factor: $$e^{ln(1 + \cos x)} = 1 + \cos x$$

Multiplying both side with I.F.

$$(1 + \cos x)^2y' - y\sin x = 2x$$

$$[(1 + \cos x)y]' = 2x$$

$$(1 + \cos x)y = x^2 + C$$

$$y = \frac{x^2 + C}{1 + \cos x}$$

Suppose initial condition $y(0) = 1$, then we can get $C$

$$1 = \frac{C}{2}$$

$$C = 2$$

$$y = \frac{x^2 + 2}{1 + \cos x}$$

## Linear with $k$ constant

$$\frac{dT}{dt} + kT = kT_e$$

I.F. $e^{kt}$, multiplied by $e^{kt}$

$$(e^{kt}T)' = kT_e e^{kt}$$

$T_e$ is the function of time t

Integrate: $$e^{kt}T = \int kT_e(t)e^{kt} + C$$

$$T = e^{-kt}\int kT_e(t)e^{kt}\ dt + Ce^{-kt}$$

Initial condition, $T(0) = T_0$

$$T = e^{-kt}\int_0^t kT_e(t_1)e^{kt_1}\ dt_1 + Ce^{-kt}$$

At $t = 0$, the integral term disappear, $C = T_0$.

As $t \to \infty$, $Ce^{-kt} \to 0$ provided $k > 0$

Hence the equation can be seen as $$T = \text{steady state solution} + \text{transient solution}$$
