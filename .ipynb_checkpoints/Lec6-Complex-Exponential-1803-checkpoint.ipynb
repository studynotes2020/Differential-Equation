## Polar form

$$a + bi = r \cos \theta + ir\sin \theta = r(\cos \theta + i\sin \theta)$$

$$e^{i\theta} = \cos \theta + i\sin \theta$$

1. This need to satisfy exponential law: $a^x.a^y = a^{x + y}$
2. $e$ is important as $e^{at}$ is the solution for $\frac{dt}{dt} = ay$ with initial condition $y(0) = 1$
3. Infinite series should work out

So in order for definition of $e^{i\theta}$ to be true, we need to show two properties

1. $e^{i\theta_1}.e{i\theta_2} = e^{i(\theta_1 + \theta_2)}$
2. $\frac{d}{dt}e^{i\theta} = ie^{i\theta}$



### Check first property

$$\begin{align}
(\cos\theta_1 + i\sin\theta_1).(\cos\theta_2 + i\sin\theta_2) & = \cos\theta_1 \cos\theta_2 - \sin\theta_1 \sin\theta_2 + i(\sin\theta_1 \cos\theta_2 + \sin\theta_2 \cos\theta_1) \\ 
& = (\cos\theta_1 + \cos\theta_2) + i(\sin\theta_1 + \sin\theta_2)
\end{align}$$

From the trigonometry identities.

In other words, the identity of $e^{i\theta}$ simplify the expression very much

### Check second property

Think of input $\theta$ with output complex value $e^{i\theta}$

In complex number general form, $u(t) + i v(t)$

$D$ for derivative. $D(u + iv) = Du + i(Dv)$.

$$\begin{align}
\frac{d}{dt}e^{it} & = \frac{d}{dt}(\cos t + i\sin t) \\
& = -\sin t + i\cos t \\
& = i(\cos t + i\sin t) \\
& = ie^{it}
\end{align}$$

What about initial condition? $e^i0 = \cos 0 + i\sin 0 = 1$

Other exponential property should also work

$$e^{a + ib} = e^a.e^{ib}$$

Polar form: $\alpha = re^{i\theta}$

$r =$ modulus of $\alpha$, $|\alpha|$

$\theta =$ argument of $\alpha$, $arg(\alpha)$

## Advantage of Polar Form

- GOOD FOR MULTIPLICATION

$$r_1e^{i\theta_1} . r_2 e^{i\theta_2} = r_1 r_2 e^{i(\theta_1 + \theta_2)}$$

Multiply the moduli, add the arguments

- Polar form also makes Calculus easy. Complexify Integral

$$\int e^{-x}\cos x \ dx = \Re (e^{-x}.e^{ix})$$

We want the real part only.

$$\int e^{-x + ix} \ dx = \frac{e^{-x + ix}}{-1 + i} = \frac{1}{-1 + i}e^{-x}(\cos x + i\sin x)$$

$$\int e^{-x + ix} \ dx = \frac{-1 - i}{2} e^{-x}(\cos x + i\sin x)$$

The real part is $\frac{e^{-x}}{2}(-\cos x + \sin x)$, hence 

$$\int e^{-x}\cos x \ dx = \frac{e^{-x}}{2}(-\cos x + \sin x)$$

### Another example of application of polar form

Calculate n-th root of 1, $\sqrt[5]{1}$. (5 answers as complex numbers)

Geometrically, we can draw unit circle (r = 1), with 5 points on the circle

$\alpha = \frac{2\pi}{5}$. It is geometrically clear that it is one of the five roots of 1. 

$$\zeta = e^{i.\frac{2\pi}{5}}$$

One can check the result by calculating power of 5

$$\zeta^5 = e^{i\frac{2\pi}{5}.5} = e^{i2\pi} = 1$$
