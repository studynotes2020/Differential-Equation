$$y'' + by' + ky = 0$$

try $y = e^{rt}$, we can see root satisfy $$r^2 + br + k = 0$$

Oscillation correspond to complex root of $r^2 + br + k = 0$

$$r = a \pm bi$$

Last lecture took $r = a + bi$, leads to solution $e^{(a + ib)t}$. We separate the exponential into real and imaginary part. Both of those give solution $y_1 = e^{at}\cos bt, y_2 = e^{at}\sin bt$

We need real solutions, because those equations are from real quantities. 

How about $r = a - bi$? That will give us 

$$y_1 = e^{at}\cos(-bt) =  e^{at}\cos bt$$
$$y_2 = e^{at}\sin (-bt) = -e^{at}\sin bt$$

The general solution is $y = c_1y_1 + c_2y_2$, so that will change the sign of $c_2$ but doesn't give any new solution


General Solution

$$y = c_1 e^{(a + bi)t} + c_2 e^{(a - bi)t}$$

We are only interested in real solution, but above is a complex function. $c_1, c_2$ are also allowed to be complex. How do I get real solution from complex function?

What are real solution? Let say the solution is $u + iv$

First way, substitute $c_1, c_2$ to be complex numbers, expand the complex exponential, collect the terms and equal the imaginary part to zero and see what it looks like. $(v = 0)$

Second way, change i to -i, see if stays the same. If after changing, the expression doesn't change, v must be zero.

Starts from here, $$y = c_1 e^{(a + bi)t} + c_2 e^{(a - bi)t}$$

Change $i$ to $-i$, and $c$ to its complex conjugate 

$$\overline y = \overline c_1 e^{(a - bi)t} + \overline c_2 e^{(a + bi)t}$$

We want these two equations to be the same. That happen when 

$$c_2 = \overline c_1, \overline c_2 = c_1$$

Complex conjugate of a complex conjugate get back the same thing. So above is only one equation. Substitute $c_1 = c + id$

$$y = (c + id) e^{(a + bi)t} + (c - id) e^{(a - bi)t}$$

Many scientists and engineers write the solution this way.



We can write the above form into sine and cosine

$$y = e^{at}[c (e^{bit} + e^{-ibt}) + id (e^{ibt} - e{-ibt})]$$

From Euler,
$$\cos a = \frac{e^{ia} + e^{-ia}}{2}$$
$$\sin a = \frac{e^{ia} - e^{-ia}}{2i}$$

$$y = e^{at}[c (2\cos bt) + id (2i\sin bt)]$$

$$y = e^{at}[2c\cos bt - 2d\sin bt]$$

## Back to this form

$$y'' + \frac{c}{m}y' + \frac{k}{m}y = 0$$

Write this in this form. So that later the equation can be written nicer.

$$y'' + 2py' + \omega_0^2y = 0$$

Oscillations (only happen when the equation has copmlex roots)

$$r^2 + 2pr + \omega_0^2 = 0$$

$$r = -p \pm \sqrt{p^2 - \omega_0^2}$$

If $p = 0$, the equation becomes $y'' + \omega_0^2 y = 0$. This is undamped case. The solutions are pure oscillation and $\omega_0$ is the circular frequency of the solutions. Because $r = \pm i\omega_0$ ($a = 0, b = \omega_0$)

$$y = c_1 \cos \omega_0t + c_2 \sin\omega_0t = A\cos(\omega_0t - \phi)$$



### Damped Case

If $p^2 - \omega_0^2 < 0 \implies p < \omega_0$, get oscillations. In another words, half the damping less than circular frequency.

Spacing between t-axis intersection is half "period", or $\pi/\omega_1$. $\omega_1$ is pseudo-frequency. 

If damping goes up, $\omega_1$ goes down. We can see from the equation below

$$r = -p \pm \sqrt{-(\omega_0^2 - p^2)} = -p \pm \sqrt{-\omega_1^2}$$

Solution: $$\begin{align}y & = e^{-pt}(c_1\cos \omega_1t + c_2\sin\omega_1t) \\
& = e^{-pt} A \cos(\omega_1t - \phi)
\end{align}$$

If the function cross t-axis at $t_1$, it crosses next time at $t_2$, hence

$$t_2 = t_1 + \frac{2\pi}{\omega_1}$$

At first cross t-axis: $$\omega_1 t_1 - \phi = \frac{\pi}{2}$$

At second cross t-axis: $$\omega_1 t_2 - \phi = \omega_1(t_1 + \frac{2\pi}{\omega_1}) - \phi = \frac{\pi}{2} + 2\pi$$

Hence $\omega_1$ is called pseudo-frequency

Let analyze what depends on what from this equation

$$y = e^{-pt} A \cos(\omega_1t - \phi)$$

$p$ - damping constant depends on ODE, $c/2m$. This tells us how much it is coming down

$\omega_1$ - depends on ODE, $\omega_1^2 = \omega_0^2 - p^2$. This is a triangle with $\omega_0$ is the hypotenuse. So it depends on damping and spring constant.

$\phi$ - Phase lag depends on initial condition

$A$ - Amplitude depends on initial condition
