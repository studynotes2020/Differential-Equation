# Autonomous ODS

$$y' = f(y)$$

No $t$ on RHS. Time invariant. This is called autonomous. No independent variable on RHS

Problem: get qualitative info about solutions

Direction field

Horizontal lines $y = y_0$. Isoclines

The slopes, $f(y_0)$ are constant

Critical Point is $y_0$ and $f(y_0) = 0$

$y = y_0$ is a solution

Proof: $$\frac{dy_0}{dt} = f(y_0)$$

Both sides is zero, so the statement is true

1. Find critical point by $f(y_0) = 0$
2. Graph $f(y)$ for greater than zero and less than zero
$\frac{dy}{dt} = f(y)$ so if $f(y) > 0$ then solution $y(t)$ must be increasing



### Example

$y = \text{money in account}$

$r = \text{constant interest rate}$

$w = \text{rate of embezzlement}$

$$\frac{dy}{dt} = ry - w$$

1. Critical points: $ry - w = 0$

$$y = \frac{w}{r}$$

2. Graph the function $\frac{dy}{dt}$ vs $y$. It has slope r, intercept vertical axis at -w. At intercept horizontal line is critical point. To the right, the functioin increasing. To the left, the function decreasing.

3. Graph $y$ vs $t$. At the region above w/r, the function slope increasing. At the region below, the function decreasing. 



## Logistic Equation

It describes population behavior $y(t)$

$$\frac{dy}{dt} = ky$$
where $k$ is the growth rate

If $k$ is constant, it is a simple growth. But the growth rate k declines as population y increases as more resources are depleted.

A simple choice: $k = a - by$

Logistic growth: $$\frac{dy}{dt} = ay - by^2$$

To solve it, separation variables and partial fractions

Step 1, find critical points. We can factor out $y$ from the equation: 

$$\frac{dy}{dt} = y(a - by) = 0$$

$y = 0, y = \frac{a}{b}$ These are two critical points

Step 2, graph $y'$ vs $y$ and then $y$ vs $t$



```julia
using Plots
f(y) = y*(4-2y)
plot(f, 0, 3)  # plot y' vs y
```

```julia
# Integrate f(y) by separation of var get 1/4*(ln(y/(2-y))) = t
# hence y = 2e^(4t)/(1 + e^(4t))
g(t) = 2*exp(4t)/(1 + exp(4t))  # plot y vs t
plot(g, -2, 2)
```

## Logictic Equation with Harvesting

Assume harvest at the constant time rate, h

$$\frac{dy}{dt} = ay - by^2 - h$$

Notice that if LHS is zero, RHS is a quadaratic solution. The quadratic solution will have two critical points if h is small. When h becomes bigger and bigger, the curve will have only one critical point. At this situation, $h_m$ is the maximum rate of harvesting. Maximum rate one can harvest salmon, without the population of salmon goes to zero

Further that, the curve will have no critical point and the population will go to zero


```julia
plot(x->(-x^2 + 3x - 2), 0, 3)  # Two critical points
plot!(x->(-x^2 + 3x - 2.25), 0, 3)  # One critical points hm is 2.25
plot!(x->(-x^2 + 3x - 3), 0, 3)  # No critical points
```
