# Autonomous Non-Linear System

$$x' = f(x, y)$$
$$y' = g(x, y)$$

Allow non-linear function on RHS (not ax + by), like quadratic, sines, cosines

Problem: Sketch its trajectories

Do it with example: Non-linear pendulum (lightly damped)

## Non-linear Pendulum

A pendulum with length l, mass m and angle theta. It satisfies Newton law

$$m\vec a = \vec F$$

The acceleration is angular acceleration and force is from gravity but we only interested to the component that causes the pendulum to swing which is equal to $mg\sin\theta$. There is also a damping force resisting the force

$$ml\theta'' = -mg\sin\theta - c_1l\theta'$$

Rearrange

$$\theta'' + \frac{c_1}{m}\theta' + \frac{g}{l}\sin\theta = 0$$

Lumping the constants and gives them new names

$$\theta'' + c\theta' + k\sin\theta = 0$$

This is second order non-linear equation

Convert into first order system

$$\theta' = w$$
$$w' = -k\sin\theta - cw$$

Take c = 1, k = 2 to make a light damped system

$$\theta' = w$$
$$w' = -2\sin\theta - w$$

How to solve?

1. Look for critical points of the system, $(x_0, y_0)$ where $f(x_0, y_0) = 0$ and $g(x_0, y_0) = 0$. Start at position zero with velocity zero. This solves the system as $x = x_0, y = y_0$, sit rest, for all time

Critical points: $w = 0 \implies \sin\theta = 0 \implies \theta = 0, \pm\pi, \pm 2\pi...$

This means pendulum starts at bottom and sitting rest. Another point is pendulum at the top and sitting rest. Physically, point $(0, 0)$ is stable critical point where $(\pi, 0)$ is unstable. Basically just two critical points.

2. For each critical point, linearize system near it, plot trajectories of this linearize system

Linearize at (0, 0). $\sin\theta \approx \theta$ when $\theta$ is small ($\theta \approx 0$)

$$\theta' = w$$
$$w' = -2\theta - w$$

$$A = \begin{bmatrix}0 & 1 \\ -2 & -1 \end{bmatrix}$$

Characteristic equation: $$\lambda^2 + \lambda + 2 = 0$$

$$\lambda = -\frac{-1 \pm \sqrt{-7}}{2}$$

Complex root means spiral sink in plot omega vs theta. Theta decreasing with decreasing angular velocity then to a point theta starts increasing with increasing angular velocity. Meanwhile, the magnitude of both theta and omega is decreasing, and that's why the curve spiral down to the origin

The plot is a description of pendulum swinging back and forth. 

Then we do the next critical point. Linearize at point $(\pi, 0)$. The theta is not small, so we cannot approximate sine to theta, instead, we should use Jacobian matrix.

Jacobian matrix: $$J = \begin{bmatrix} f_x & f_y \\ g_x & g_y \end{bmatrix}$$

Evaluate at $(x_0, y_0)$. We will get the matrix of the linearized system at that point

The system is 
$$\theta' = w$$
$$w' = -2\sin\theta - w$$

Here x is theta, y is omega

$$J = \begin{bmatrix} 0 & 1 \\ -2\cos\theta & -1 \end{bmatrix}$$
 
Check with point (0, 0) which we have done before.
 
$$J = \begin{bmatrix} 0 & 1 \\ -2 & -1 \end{bmatrix} = A$$

It is the same. Now at point $(\pi, 0)$

$$J = \begin{bmatrix} 0 & 1 \\ 2 & -1 \end{bmatrix}$$

Characteristic equation: $$\lambda^2 + \lambda - 2 = 0$$

$$(\lambda + 2)(\lambda - 1) = 0$$

$\lambda = 1 \implies -a_1 + a_2 = 0 \implies \begin{bmatrix} 1 \\ 1\end{bmatrix} e^t$

$\lambda = -2 \implies 2a_1 + a_2 = 0 \implies \begin{bmatrix} 1 \\ -2\end{bmatrix} e^{-2t}$

At $(\pi, 0)$, it is a saddle. If theta a little less than pi, the pendulum swing clockwise, omega negative and getting much more negative, and theta decreases also. If theta a little more than pi, the pendulum swing anti-clockwise, theta now increases and get to $2\pi$ and continue to oscillate around $2\pi$. Meanwhile omega positive and getting more positive.

3. Big picture. Plot trajectories around each critical point, connects them

Graph omega vs theta

At $\theta = 0, 2\pi, 4\pi,...$, a spiral curve, spinning clockwise to the critical points

At $\theta = 0, \pi, 3\pi,...$, a saddle curve.
