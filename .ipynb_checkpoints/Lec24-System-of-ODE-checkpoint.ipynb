# Systems

One independent variable, many dependent variables

$$x' = f(x, y; t)$$
$$y' = g(x, y; t)$$

x, y are dependent variables, depend on independent variable t. $x' = \frac{dx}{dt}, y' = \frac{dy}{dt}$


## Linear System

$$x' = ax + by + r_1(t)$$
$$y' = cx + dy + r_2(t)$$

a, b, c, d are constant, then it is called constant-coefficients system. 

a, b, c, d can be also function of t

First order means highest derivative is one

Homogeneous means $r_1(t), r_2(t)$ is zero

Number of Initial condition is the total order of the system. In this example, it is two. $x(t_0) = x_0, y(t_0) = y_0$

Suppose a circuit with resistor, capacitor and source, then it is a first-order equation. If it has two loops, with an additional capacitors and resistor, then there is a pair of equations. Suppose there is a third loop with inductor, then it is system of three equations, first order, first order and second order


### Example

Suppose there is an egg inside a tank. The outer temp is $T_e(t)$, egg white has temperature $T_2$ and york has temperature $T_1$

$$\frac{dT_1}{dt} = a(T_2 - T_1)$$
$$\frac{dT_2}{dt} = a(T_1 - T_2) + b(T_e(t) - T_2)$$

Better version

$$T_1' = -aT_1 + aT_2$$
$$T_2' = aT_1 - (a + b)T_2 + bT_e(T)$$

$T_e$ is not variable but some function of t. If coddling, it might be $T_e(t) = 100e^{-kt}$

If in icebath, $T_e(t) = 0$, a homegeneous system

Let the system in icebath and a = 2, b = 3

$$T_1' = -2T_1 + 2T_2$$
$$T_2' = 2T_1 - 5T_2$$

$$T_2 = \frac{T_1' + 2T_1}{2}$$

substitute into (2)

$$(\frac{T_1' + 2T_1}{2})' = 2T_1 - 5( \frac{T_1' + 2T_1}{2})$$

Two first order, produce a second order equation. The sum of order is the same.

$$T_1'' + 7T_1' + 6T_1 = 0$$

Characteristic equation

$$r^2 + 7r + 6 = 0$$

$$(r + 1)(r + 6) = 0$$

$r = -1, r = -6$ two real and distinct roots. So the solution is

$$T_1 = c_1e^{-t} + c_2e^{-6t}$$

But we shouldn't go through the same process to get $T_2$ which will give us two more unknowns, instead, we should get $T_2$ from the solution of $T_1$

$$T_2 = \frac{T_1' + 2T_1}{2}$$

$$T_2 = \frac{c_1}{2}e^{-t} - 2c_2e^{-6t}$$

To find $c_1, c_2$, let say $T_1(0) = 40, T_2(0) = 45$

$$40 = c_1 + c_2$$
$$45 = c_1/2 - 2c_2$$

$c_1 = 50, c_2 = -10$

$$T_1 = 50e^{-t} - 10e^{-6t}$$
$$T_2 = 25e^{-t} + 20e^{-6t}$$

The solution tells us how the temperature of york and white varies with time

## Autonomous System

$$x' = f(x, y)$$
$$y' = g(x, y)$$

No t appears no the RHS. Autonomous system. Not a linear system.

Solution: $x = x(t), y = y(t)$ which is a parameterized curve

What does x' and y' means?

Velocity vector. Velocity of the solution at time t. Its graph is called velocity field

A system of first order equation ODE (autonomous) = velocity field

Solution is parameterized curve with the right velocity everywhere

There is an important connection between velocity field and direction field taught in the first lecture.
