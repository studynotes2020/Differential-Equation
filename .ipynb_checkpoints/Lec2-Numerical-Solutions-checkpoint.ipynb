# Numerical Solutions

Computer solves the differential equations with numerical solutions

IVP, Initial Value Problem

$$y' = f(x, y) \quad y(x_0) = y_0$$


## Euler Method

A point $(x_0, y_0)$. Choose a step size, $h$, calculate slope at that points then connects them

If step size is $h = x_{n+1} - x_n$, slope is $A_n$, then height of the triangle is $hA_n = y_{n+1} - y_n$

$$y_{n+1} = hA_n + y_n$$


## Euler Equations

$$x_{n+1} = x_n + h$$

$$y_{n+1} = y_n + hA_n$$

$$A_n = f(x_n, y_n)$$ 


### Example

An equation cannot be solved by elementary function, so this is a good example for numerical method

$$y' = x^2 - y^2, \quad y(0) = 1, \quad h = 0.1$$

$n = 0 \implies x_n = 0, y_n = 1, A_n = -1, hA_n = -.1$

$n = 1 \implies x_n = .1, y_n = .9, A_n = -.80, hA_n = -.08$

$n = 2 \implies x_n = .2, y_n = .82, A_n =, hA_n = $

$y_2(.2) \approx .82$ Is Euler too high or too low?

If the curve is convex, $y''> 0 $, Euler is too low. 

If concave, Euler is too high, $y''< 0 $

$$y' = x^2 - y^2$$

$$y'' = 2x - 2yy'$$

At $(0, 1), \quad y(0) = 1, \quad y'(0) = -1, \quad y''(0) = 2(0) - 2(1)(-1) = 2$ convex, so Euler too low




## Better method

Use a smaller step size. Error e will be smaller. e depends on step size

$e ~ c_1h$, Euler is a first-order method

If halve the step size, halve the error. 


## Better method

Find a better slope

Let first slope called $A_n$, second slope $B_n$, then take the average of $A_n, B_n$

$x_{n+1} = x_n + h$

$y_{n+1} = y_n + h(\frac{A_n + B_n}{2}) \quad B_n = f(x_{n+1}, \tilde y_{n+1}) \quad \tilde y_{n+1} = y_n + hA_n$

$B_n, \tilde y$ is just temperory for calculating $y_{n+1}$

Four names for this method 

- Heun's method
- Improved Euler
- Modified Eulter
- RK2 - Runge-Kutta

This is a second-order method. $e ~ c_2h^2$, halve the step size, error goes down a quarter

There is an RK4, it is a standard method, accurate. Requires to calculate four slopes, $\frac{A_n + 2B_n + 2C_n + D_n}{6}$
