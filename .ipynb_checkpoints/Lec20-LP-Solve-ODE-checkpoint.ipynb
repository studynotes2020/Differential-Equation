# Using LP to Solve ODE

Laplace Transform: $$F(s) = \int_0^\infty e^{-st}f(t)\ dt$$

The integration cannot "blow up", or should converge to a value. In order for that to be true, $f(t)$ should be an "exponential type"

If $f(t)$ is of "exponential type", then $|f(t)| \leq Ce^{kt}$

$C$ and $k$ is some positive constant. This is true for t >= 0. We don't care about t < 0 because Laplace tranform only intergrate from 0 to infinity

1. $f(t) = \sin t$ is an example of exponential type as $|\sin t| \leq 1e^{0t}$

2. $f(t) = t^n$. $t^n \leq Me^t$ for some M, n >=0 and all t > 0. $t^n$ is of exponential type

$$\frac{t^n}{e^t} \leq M ?$$

By L'Hospital Rule n times: $$\frac{t^n}{e^t} \to 0$$ as $t \to \infty$. Because we after n derivation, numerator becomes $n!$, and the denominator is $t^ne^t$ which is clearly bigger than numerator, os limit approaches zero

## What function is not exponential type?

$$f(t) = \frac{1}{t}$$

$$\int_0^\infty e^{-st}\frac{1}{t}\ dt$$ doesn't converge.

So 1/t is not an exponential type and doesn't has a LP

Another function $f(t) = e^{t^2}$ is not of exponential type, because it grows too fast

$$e^{t^2} > e^{kt}$$

If $t^2 > kt$, then $t > k$. So $e^{t^2}$ also doesn't has a LP

## Initial Value Problem

$$y'' + Ay' + By = h(t)$$

LP must has initial condition in order to solve ODE. If not given, just gives it a variable. $y(0) = y_0$ and $y'(0) = y_0'$

Laplace transform the ODE into algebraic equation of $Y(s)$. Solve it for $Y(s)$. Usually in the form $$Y(s) = \frac{p(s)}{q(s)}$$

Finally inverse LP $Y(s)$ into $y(t)$

The reason of doing this indirect route because going from ODE to y(t) is very hard but going from ODE to Y(s) is often easy. The inverse LP is usually hardest in the LP method

Starts from $f'(t)$, so that after we do integration by parts, we get the expression we want

$$\begin{align}\mathscr{L}(f'(t)) & = \int_0^\infty e^{-st}f'(t)\ dt \\
& = e^{-st}f(t)|_0^\infty - \int_0^\infty -se^{-st} f(t)\ dt
\end{align}$$

We want to know the limit with s > 0 $$\lim \limits_{t \to \infty} \frac{f(t)}{e^{st}}$$ 

Since $f(t)$ is of exponential type

$$\lim \limits_{t \to \infty} \frac{f(t)}{e^{st}} = 0$$ 

From the result of integration by parts

$$e^{-st}f(t)|_0^\infty = 0 - f(0)$$

$f(0)$ appears here, that why we need initial condition. The next term

$$\int_0^\infty se^{-st} f(t)\ dt = sF(s)$$

Collect the term and assuming $f'(t)$ is of exponential type

$$\mathscr{L}(f'(t)) = sF(s) - f(0)$$

How about second derivative?

$$\begin{align}\mathscr{L}(f''(t)) & = \mathscr{L}([f'(t)]') \\ 
& = s\mathscr{L}(f'(t)) - f'(0) \\ 
& = s[s(F(s) - f(0)] - f'(0)
\end{align}$$

$$\mathscr{L}(f''(t)) = s^2F(s) - sf(0) - f'(0)$$

### Example

$$y'' - y = e^{-t}$$ $y(0) = 1, y'(0) = 0$

Laplace tranform the equation

$$s^2Y - s - Y = \frac{1}{s + 1}$$

$$(s^2 - 1)Y = \frac{1}{s + 1} + s = \frac{s^2 + s + 1}{s + 1}$$

$$Y = \frac{s^2 + s + 1}{(s + 1)^2 (s - 1)}$$

We want to use partial fraction, so that it can be applied inverse Laplace transform

$$\frac{s^2 + s + 1}{(s + 1)^2 (s - 1)} = \frac{-1/2}{(s + 1)^2} + \frac{1/4}{s + 1} + \frac{3/4}{s - 1}$$

Separate one more term, 1/(s + 1), so that all the numerator is constant

$$\mathscr{L^{-1}}(Y) = -\frac{1}{2}te^{-t}+ \frac{1}{4}e^{-t} + \frac{3}{4}e^t$$

The first term is the particular solution. Second and third term is the complementary function
