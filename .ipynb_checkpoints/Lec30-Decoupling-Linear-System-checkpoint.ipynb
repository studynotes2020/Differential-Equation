## Methods Solving Homogeneous Linear System

$$\vec x' = A\vec x$$

Methods to solve

1. Eigenvalues, eigenvectors


2. $e^{At}\vec x_0$. Use eigenvalues and eigenvectors to find $X$, then get the fundamental matrix by $e^{At} = X(t)X(0)^{-1}$


3. Decoupling: introduce new variables u, v so that the system decoupled

$$u = ax + by$$
$$v = cx + dy$$

Decoupled: 
$$u' = k_1u$$
$$v' = k_2v$$

## Solve with Decoupling

Suppose a tank with two chambers, one with height x, another with height y. The area of second chamber is twice the first one. There is a hole at the bottom that allows liquid to flow between each other. The flow rate is 2. We know that in the end the height x will be equal to height y.

$$x' = 2(y - x)$$
$$2y' = 2(x - y)$$

$$x' = -2x + 2y$$
$$y' = x - y$$

We want to introduce new variables so that the system becomes two single equation. Let u be total amount of water and v be the pressure on the hole

$$u = x + 2y$$
$$v = x - y$$

Then we get new system in terms of u and v

$$u' = x' + 2y' = -2x + 2y + 2x - 2y = 0$$
$$v' = x' - y' = -2x + 2y - x + y = -3x + 3y$$

$$u' = 0$$
$$v' = -3v$$

The system is decoupled

Solutions: 
$$u = c_1$$
$$v = c_2e^{-3t}$$

In terms of x and y
$$x = \frac13 (u + 2v) = \frac13 (c_1 + 2c_2e^{-3t})$$
$$y = \frac13 (u - v) = \frac13 (c_1 - c_2e^{-3t})$$

$$\vec x = \frac13 c_1 \begin{bmatrix}1 \\ 1 \end{bmatrix} + \frac13 c_2 \begin{bmatrix}2 \\ -1 \end{bmatrix} e^{-3t}$$

Notes
1. There is a general method of choosing new variables
2. We cannot always decouple. In order to decouple, the eigenvalues must be real, complete (non-defective). A real and symmetric matrix is always complete and can be decoupled





## Decoupling Matrix

We start with this
$$\begin{bmatrix}u \\ v \end{bmatrix} = 
\begin{bmatrix}a_1' & a_2' \\ b_1' & b_2' \end{bmatrix}
\begin{bmatrix}x \\ y \end{bmatrix}$$

$$\vec u = D\vec x$$

Decoupling matrix
$$D = \begin{bmatrix}a_1' & a_2' \\ b_1' & b_2' \end{bmatrix}$$

But what we need is to get x and y in terms of u and v in order to do substitution

$$\vec x = E\vec u$$

$$D^{-1} = E = \begin{bmatrix}a_1 & a_2 \\ b_1 & b_2 \end{bmatrix}$$

$$\begin{bmatrix}x \\ y \end{bmatrix} = 
\begin{bmatrix}a_1 & a_2 \\ b_1 & b_2 \end{bmatrix}
\begin{bmatrix}u \\ v \end{bmatrix}$$

Columns are the two eigenvectors

$$E = \begin{bmatrix}\vec a_1 & \vec a_2 \end{bmatrix}$$

Then substitute above equation into $\vec x' = A\vec x$ to see if it is decoupled in uv coordinate. But before that, we need to use the result of $AE$

$$\begin{align} AE & = A\begin{bmatrix}\vec a_1 & \vec a_2 \end{bmatrix} \\
& = \begin{bmatrix}A\vec a_1 & A\vec a_2 \end{bmatrix} \\
& = \begin{bmatrix}\lambda_1 \vec a_1 & \lambda_2 \vec a_2 \end{bmatrix} \\
& = \begin{bmatrix}\vec a_1 & \vec a_2 \end{bmatrix}
\begin{bmatrix}\lambda_1 & 0 \\ 0 & \lambda_2 \end{bmatrix}
\end{align}$$

Knowing that $A\vec a = \lambda \vec a$ where $\lambda, \vec a$ is eigenvalue and eigenvector

$$AE = E\begin{bmatrix}\lambda_1 & 0 \\ 0 & \lambda_2 \end{bmatrix}$$

Substitute $\vec x = E\vec u$ into $\vec x' = A\vec x$ to see if it is decoupled in uv coordinate

$$Eu' = AE\vec u = E\begin{bmatrix}\lambda_1 & 0 \\ 0 & \lambda_2 \end{bmatrix} \vec u$$

$$\vec u' = \begin{bmatrix}\lambda_1 & 0 \\ 0 & \lambda_2 \end{bmatrix} \vec u$$

Write the system out

$$u' = \lambda_1 u$$
$$v' = \lambda_2 v$$

The system is decoupled

Solutions:
$$u = c_1 e^{\lambda_1 t}$$
$$v = c_2 e^{\lambda_2 t}$$

## General Method with Decoupling Matrix

$$\begin{bmatrix}x \\ y \end{bmatrix}' = 
\begin{bmatrix}-2 & 2 \\ 1 & -1 \end{bmatrix} 
\begin{bmatrix}x \\ y \end{bmatrix}$$

Calculate eigenvalues and eigenvectors. First the characteristic equation is

$$\lambda^2 + 3\lambda = 0$$

$\lambda = 0, -3$

$\lambda = 0 \implies -2a_1 + 2b_1 = 0 \implies \vec a_1 = \begin{bmatrix}1 \\ 1 \end{bmatrix}$

$\lambda = -3 \implies a_1 + 2b_1 = 0 \implies \vec a_2 = \begin{bmatrix}-2 \\ 1 \end{bmatrix}$

$$E = \begin{bmatrix}1 & -2 \\ 1 & 1 \end{bmatrix}$$

$$D = E^{-1} = \frac13 \begin{bmatrix}1 & 2 \\ -1 & 1 \end{bmatrix}$$

$$\begin{bmatrix}u \\ v \end{bmatrix} = 
\frac13 \begin{bmatrix}1 & 2 \\ -1 & 1 \end{bmatrix}
\begin{bmatrix}x \\ y \end{bmatrix}$$

$$u = \frac13 (x + 2y), v = \frac13 (-x + y)$$

Only differs with the previous only with a constant $\frac13$, but it doesn't matter because both satisfy the differential equation $u' = 0, v' = -3v$
