# Jump Discontinuities

Example: Unit step function, $u(t)$. $u(0)$ is undefined. $u_a(t) = u(t - a)$

Another example: Unit box function, $u_{ab}(t)$. The function rise to 1 at a, then goes back zero at b. Write it in terms of unit step function, which is more convenient in taking LP

$$u_{ab}(t) = u_a(t) - u_b(t) = u(t - a) - u(t - b)$$

In effect of multiplying an arbitrary function with unit box function, we wipe away all of the graph, except the part between a and b. This is very useful in the application

$$L(u(t)) = \int_0^\infty e^{-st}\ dt = \frac{1}{s}, s > 0$$

Laplace transform of 1 is also 1/s. So when we are taking inverse Laplace transform, what is the result we should pick?

Because LP just taking integration starts from 0, so any curse in the negative region doesn't matter. So many functions have the same Laplace transform. LP only uses for problem to the future, no need to have any info about the past. If need to know about the past, we need Fourier transform

To make inverse LP unique, we make any function in negative region zero by multiplying unit step function

$$L^{-1}(F(s)) = u(t)f(t)$$ 


## Want formula 

$L(f(t - a)$ in terms of $L(f(t)) \to$ doesn't exist


```julia
using Plots
plot(x->x^2, 0, 4)
plot!(x->(x-2)^2, 0, 4)
```

From the graph above, y2 or f(t-a) is a translation of y1 or f(t) to the right. The curve of f(t-a) between 0-2 doesn't get used in doing LP for f(t). But it needed to do LP for f(t-a). The information is lost in L(f(t)), so L(f(t-a)) can't be expressed in term of L(f(t))

Right formula for $$L(u(t-a)f(t-a)) = e^{-as}F(s)$$

This is like dual to the previous formual where $e^{-as}$ on the left and $F(s-a)$ on the right. And the above one is $e^{-as}$ on the right and $f(t-a)$ on the left

Another formula we are going to prove is t-axis translation formula 
$$L(u(t-a)f(t)) = e^{-as}L(f(t+a))$$

To prove the LP for the first one. Starts with LP definition and change the variable. This is called changing variable in the definite integral. $t_1 = t - a$. When t is zero, $t_1 = -a$. 

$$\begin{align}\int_0^\infty e^{-st}u(t-a)f(t-a)\ dt & = \int_{-a}^\infty e^{-s(t_1+a)}u(t_1)f(t_1)\ dt_1 \\
& = e^{-as}\int_{-a}^\infty e^{-st_1}u(t_1)f(t_1)\ dt_1 \\
& = e^{-as}\int_{0}^\infty e^{-st_1} f(t_1)\ dt_1 \\
& = e^{-as}F(s)
\end{align}$$

In the range $t = [0, \infty], u(t_1) = 1$ or another words $u(t_1) = 0$ for $t < 0$ so we can change lower limit of integral from -a to 0, and take out $u(t_1)$

Hence, we prove that $$L(u(t-a)f(t-a)) = e^{-as}F(s)$$

We can also think as $$L(u(t-a)f(t-a)) = e^{-as}L(f(t))$$

To prove the second formula, we slightly change the expression

$$u(t-a)f(t) = u(t-a)f(t-a+a)$$

Then its LP is just replace t by t + a at RHS

$$L(u(t-a)f(t-a+a)) = e^{-as}L(f(t+a))$$

## LP for Unit Box Function

Unit box function
$$u_{ab}(t) =  u(t - a) - u(t - b)$$

$$L(u_{ab}(t)) = \frac{e^{-as}}{s} - \frac{e^{-bs}}{s}$$

$$L(u(t-1)t^2) = e^{-s}L((t+1)^2) = e^{-s}L(t^2 + 2t + 1)$$

$$L(u(t-1)t^2) = e^{-s}(\frac{2}{s^3} + \frac{2}{s^2} + \frac{1}{s})$$

$e^{-s}$ tells us that discontinuity occurs at 1

## Inverse LP

Calculate inverse LP of the expression

$$\frac{1 + e^{-\pi s}}{s^2 + 1} = \frac{1}{s^2 + 1} + \frac{e^{-\pi s}}{s^2 + 1}$$

$$L^{-1}(\frac{1}{s^2 + 1}) = u(t)\sin t$$

$$L^{-1}(\frac{e^{-\pi s}}{s^2 + 1}) = u(t-\pi)\sin (t-\pi)$$

$$L^{-1}(\frac{1 + e^{-\pi s}}{s^2 + 1}) =  u(t)\sin t + u(t-\pi)\sin (t-\pi)$$

The first part is non-zero when t >= 0 while the second part is non-zero when t >= pi

$$f(t) =
\begin{cases}
\sin t,  0 \leq t \leq \pi \\
\sin t + (-\sin (t-\pi)), t \geq \pi \\
\end{cases}$$

$$f(t) =
\begin{cases}
\sin t,  0 \leq t \leq \pi \\
0, t \geq \pi \\
\end{cases}$$
