# Solve by Matrix Method

From Last Lecture

$$ x' = -2x + 2y$$
$$ y' = 2x - 5y$$

Solved by elimination last time, we got

$$x = c_1e^{-t} + c_2e^{-6t}$$
$$y = \frac{c_1}{2}e^{-t} - 2c_2e^{-6t}$$

If transform the problem and solution into matrix form

$$\begin{bmatrix} x \\ y \end{bmatrix}' = 
\begin{bmatrix} -2 & 2 \\ 2 & -5 \end{bmatrix}
\begin{bmatrix} x \\ y \end{bmatrix}$$

$$\begin{bmatrix} x \\ y \end{bmatrix} = 
c_1\begin{bmatrix} 1 \\ \frac12 \end{bmatrix} e^{-t} + 
c_2\begin{bmatrix} 1 \\ -2 \end{bmatrix} e^{-6t}$$

Trial solution

$$\begin{bmatrix} x \\ y \end{bmatrix} = 
\begin{bmatrix} a_1 \\ a_2 \end{bmatrix}e^{\lambda t}$$

$$\begin{bmatrix} x \\ y \end{bmatrix}' = 
\lambda \begin{bmatrix} a_1 \\ a_2 \end{bmatrix}e^{\lambda t} = 
\begin{bmatrix} -2 & 2 \\ 2 & -5 \end{bmatrix}
\begin{bmatrix} a_1 \\ a_2 \end{bmatrix}e^{\lambda t}$$

$e^{\lambda t}$ can be cancelled and this substitution reduces differential equation into algebraic equation to solve for $\lambda, a_1, a_2$

$$\lambda a_1 = -2a_1 + 2a_2$$
$$\lambda a_2 = 2a_1 - 5a_2$$

Two equations with three unknowns.

$$(-2-\lambda)a_1 + 2a_2 = 0$$
$$2a_1 + (-5-\lambda)a_2 = 0$$

It has a trivial solution of $a_1 = 0, a_2 = 0$ but we don't want that solution. We want non-trivial solution. It happens when matrix on LHS is invertible or the determinant is zero

$$\begin{vmatrix} -2-\lambda & 2 \\ 2 & -5-\lambda \end{vmatrix} = 0$$

$$(\lambda + 2)(\lambda + 5) - 4 = 0$$

$$\lambda^2 + 7\lambda + 6 = 0$$

This is characteristic equation with roots $\lambda = -1, \lambda = -6$

For each value of $\lambda$ we need to find corresponding value of $a_1, a_2$

When $\lambda = -1$

$$-a_1 + 2a_2 = 0$$
$$2a_1 - 4a_2 = 0$$

The second equation is a constant multiple of the first one. This is the consequence of finding lambda such that the system would be redundant, therefore will have non-trivial solution. By letting $a_2 = 1$, we have $a_1 = 2$

Solution: $$\begin{bmatrix} 2 \\ 1 \end{bmatrix} e^{-t}$$

When $\lambda = -6$

$$4a_1 + 2a_2 = 0$$
$$2a_1 + a_2 = 0$$

$a_1 = 1, a_2 = -2$

Solution: $$\begin{bmatrix} 1 \\ -2 \end{bmatrix} e^{-6t}$$

By Superposition principle, solution is $$\begin{bmatrix} x \\ y \end{bmatrix} = k_1 \begin{bmatrix} 2 \\ 1 \end{bmatrix} e^{-t} + k_2 \begin{bmatrix} 1 \\ -2 \end{bmatrix} e^{-6t}$$

## In general

$$\begin{bmatrix} x \\ y \end{bmatrix}' = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix}$$

Trial: $$\begin{bmatrix} a_1 \\ a_2 \end{bmatrix} e^{\lambda t}$$

Substitution:

$$\lambda \begin{bmatrix} a_1 \\ a_2 \end{bmatrix} = 
\begin{bmatrix} a & b \\ c & d \end{bmatrix}
\begin{bmatrix} a_1 \\ a_2 \end{bmatrix}$$

Homogeneous system:

$$\begin{bmatrix} a - \lambda & b \\ c & d - \lambda \end{bmatrix}
\begin{bmatrix} a_1 \\ a_2 \end{bmatrix} = 0$$

The system is solvable non-trivially iff its determinant is zero

$$\begin{vmatrix} a - \lambda & b \\ c & d - \lambda \end{vmatrix} = 0$$

$$(a - \lambda)(d - \lambda) - bc = 0$$

$$\lambda^2 - (a + d)\lambda + ad - bc = 0$$

$ad - bc$ is its determinant. $(a + d)$ is its trace, sum of its diagonal entries

This is the characteristic equation of the matrix A

roots $\lambda_1, \lambda_2$ is real and distinct, assume. Those are called eigenvalues of A. (also called characteristic values, proper values)

The last step, for each $\lambda_i$, find associated eigenvector $\vec a_i = \begin{bmatrix}a_{1i} \\ a_{2i} \end{bmatrix}$ by solving system 

$$\begin{bmatrix} a - \lambda & b \\ c & d - \lambda \end{bmatrix}
\begin{bmatrix} a_1 \\ a_2 \end{bmatrix} = 
\begin{bmatrix} 0 \\ 0 \end{bmatrix}$$

General solution:

$$\begin{bmatrix} x \\ y \end{bmatrix} = c_1 \begin{bmatrix} a_{11} \\ a_{21} \end{bmatrix} e^{\lambda_1 t} + c_2 \begin{bmatrix} a_{12} \\ a_{12} \end{bmatrix} e^{\lambda_2 t}$$

# True for n x n System

Rewrite the step in condensed form

$$\vec x = \begin{bmatrix} x \\ y \end{bmatrix}$$

$$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$$

$$\vec x' = A\vec x$$

Trial solution: $$\vec x = \vec a e^{\lambda t}$$

$\vec a = \begin{bmatrix} a_1 \\ a_2 \end{bmatrix}$

$$\lambda \vec a e^{\lambda t} = A\vec a e^{\lambda t}$$

$$A\vec a = \lambda \vec a$$

$$(A - \lambda I)\vec a = 0$$

$\lambda$ is a number not matrix, so to make it a matrix, multiply it with $I$. Its determinant is zero.

$$|A - \lambda I| = 0$$

to find the eigenvalues and then calculate the corresponding eigenvectors
