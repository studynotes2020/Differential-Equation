# Fourier Series

We have studied constant coefficient, second-order equation

$$y'' + ay' + by = f(t)$$

$f(t)$ is the input, the response or solution is $y(t)$

The input, $f(t)$ are extremly special: $e^t, \sin t \cos t$ 

Any periodic (period 2$\pi$, continuous function, can be represented as an infinite sum of sines and cosines

$$f(t) = c_0 + \sum_{n=1}^\infty [a_n\cos nt + b_n \sin nt]$$

n is frequency

Why does that solve the problem of general inputs for periodic functions?

Input $\to$ Response

$b_n \sin nt \to b_n y_n^{(s)}(t)$ s to indicates reponse from sines

$a_n \cos nt \to a_n y_n^{(c)}(t)$ c to indicates response from cosines

There is a constant in front, because $\cos 0t = 1$

We can have coefficient in front and do superposition, because ODE is linear. So taking sum of the input function (left side)

$$f(t) = b_n \sin nt + a_n \cos nt = \sum_1^\infty (b_n y_n^{(s)}(t) + y_n^{(c)}(t)) + c_1$$

So given periodic function $f(t)$, we can find its response by the finding its Fourier Series.


# Given $f(t)$, 2$\pi$ a period , find its Fourier Series

In other words, calculate coefficients $a_n$ and $b_n$

The calculation depends on orthogonality relation

$u(t), v(t)$ are continuous functions on $\mathbb{R}$. (Say $2\pi$ is a period)

Orthogonal on $[-\pi, \pi]$ if 

$$\int_{-\pi}^{\pi} u(t) v(t) dt = 0$$

Theorem:

$$\sin nt, n = 1,..., \infty$$
$$\cos mt, m = 0,..., \infty$$

Any two distinct ones (above) are orthogonal on interval $[-\pi, \pi]$

In other words,
$$\int_{-\pi}^{\pi} \sin 3t \cos 4t dt = 0$$

$$\int_{-\pi}^{\pi} \sin 3t \sin 60t dt = 0$$

Same thing with two cosines, or a sine and a cosine.

$$\int_{-\pi}^{\pi} \cos nt \sin nt dt = 0$$

The only thing not zero is when $u(t) = v(t)$ because it will be $u^2(t)$ which cannot be zero

If not answer zero, then what are they? (Just giving the result here)

$$\int_{-\pi}^{\pi} \sin ^2t dt = \pi$$

$$\int_{-\pi}^{\pi} \cos ^2t dt = \pi$$

## Three Ways to to Orthogonality Relation

1. Trigonometry Identities
2. Complex Exponential
3. Use ODE -> Shown here

Proof:

$m \neq n$, $\sin nt, \cos nt$ satisfy

$$u'' = -n^2u$$

This is all we need to know to prove the orthogonality relation

Let $u_n, v_m, m \neq n$ be any two of the functions, integrate their product over the interval $[-\pi, \pi]$ by part

$$\int_{-\pi}^{\pi} u_n'' v_m dt = u'_n v_m |_{-\pi}^{\pi} - \int_{-\pi}^{\pi} u'_n v'_m dt$$

$$\int_{-\pi}^{\pi} u_n'' v_m dt = -\int_{-\pi}^{\pi} u'_n v'_m dt --- (1)$$ 

$u'_n v_m |_{-\pi}^{\pi} $ is always zero, no matter u is sine or cosine, or v is sine or cosine

If v is sines, ines is zero at $\pi$ and $-\pi$. 

If u is cosine. Differentiate u become sines and is zero too at $\pi$ and $-\pi$. 

If u is sine, v is cosine, $\cos nt \cos mt$ is zero too because cosine is even function, and have same value at $-\pi$ and $\pi$

We can also integrate $v_m'' u_n$ by parts and get the same result

$$\int_{-\pi}^{\pi} v_m'' u_n dt = v'_m u_n |_{-\pi}^{\pi} - \int_{-\pi}^{\pi} v'_m u'_n dt$$

$$\int_{-\pi}^{\pi} v_m'' u_n dt = - \int_{-\pi}^{\pi} v'_m u'_n dt --- (2)$$

Also, from the relation we stated on the start of the proof, $u'' = -n^2u$

$$\int_{-\pi}^{\pi} u_n'' v_m dt = -n^2 \int_{-\pi}^{\pi} u_n v_m dt --- (3)$$

Similarly substitute v for $v'' = -n^2v$


$$\int_{-\pi}^{\pi} v_m'' u_n dt = -m^2 \int_{-\pi}^{\pi} u_n v_m dt --- (4)$$

(1), (2), (3), (4) are all equal, the only possibility with $m \neq n$ is

$$\int_{-\pi}^{\pi} u_n v_m dt = 0$$





## Given $f(t), 2\pi$ period, find $a_n, b_n$

From Fourier Series,

$f(t) = c_0... + a_k \cos kt + ... + a_n \cos nt + ... + b_n \sin nt$

$a_n \cos nt$ is the term we want. Multiply all terms with $\cos nt$

$f(t) \cos nt = ... + a_k \cos kt \cos nt + ... + a_n \cos nt + ... + b_n \sin nt \cos nt$

Integrate all between $[-\pi, \pi]$

$\int_{-\pi}^{\pi} f(t) \cos nt = ... + \int_{-\pi}^{\pi} a_k \cos kt \cos nt + ... + \int_{-\pi}^{\pi} a_n \cos^2 nt + ... + \int_{-\pi}^{\pi} b_n \sin nt \cos nt$

All other terms become zero, except the term we want

$$\int_{-\pi}^{\pi} f(t) \cos nt = \int_{-\pi}^{\pi} a_n \cos^2 nt$$

$$\int_{-\pi}^{\pi} f(t) \cos nt = a_n \pi$$

$$a_n = \frac{1}{\pi}\int_{-\pi}^{\pi} f(t) \cos nt dt$$

$n = 1, 2, ...$

The procedure to get $b_n$ is exactly the same. Look for the term we want, multiply whole equation with that term and integrate each term between $[-\pi, \pi]$. Integrate of $\sin^2 nt$ is also $\pi$

$$b_n = \frac{1}{\pi}\int_{-\pi}^{\pi} f(t) \sin nt dt$$

$n = 1, 2, ...$

But the constant term is different

$\int_{-\pi}^{\pi}f(t) = \int_{-\pi}^{\pi}c_0 + ... + \int_{-\pi}^{\pi}a_n \cos nt + ... + \int_{-\pi}^{\pi}b_n \sin nt$ 

$$\int_{-\pi}^{\pi}f(t) dt = 2\pi c_0$$

Because integration of cosine and sine is 0 over a period

$$c_0 = \frac{1}{2\pi}\int_{-\pi}^{\pi}f(t) dt$$

$c_0$ is just $a_0$ when n is zero, so we can write $c_0$ in term of $a_n$

$$f(t) = \frac{a_n}{2} + \sum_{n=1}^\infty [a_n\cos nt + b_n \sin nt]$$



## Example



```julia
using Plots
# Unit step function from -pi to pi
plot(x -> -1, -pi, 0)
plot!(x -> 1, 0, pi)
```

The function above can be expressed in Fourier Series. Because it is an odd function

$$a_n = 0$$

$$b_n = \int_{-\pi}^{0}\sin nt dt + \int_{0}^{\pi}\sin nt dt$$

$$b_n = \frac{1 - \cos n\pi}{n} - \frac{1 - \cos n\pi}{n}$$

$$b_n = \frac{2}{n}(1 - \cos n\pi)$$

$$= \frac{2}{n}(1-(-1)^n) $$

When n is odd, $b_n = \frac4n$. When n is even $b_n = 0$

$$c_0 = \frac{a_n}{2} = 0$$

After we got $c_0, a_n, b_n$ substitue back into Fourier Series

$$f(t) = \frac{a_n}{2} + \sum_{n=1}^\infty [a_n\cos nt + b_n \sin nt]$$

$$f(t) = \sum_{n=odd}^\infty \frac4n \sin nt$$

```julia
f(x) = 4 * sin(x) + 4/3*sin(3x) + 4/5*sin(5x) + 4/7*sin(7x) + 4/9*sin(9x)
plot(f, -pi, 4pi)
```
