# Inhomogeneous Equation

$$y'' + p(x)y' + q(x)y = f(x)$$

$x$ often is time. $f(x)$ can be called input, signal, driving term, forcing term

The solution $y(x)$ is response or output

The associated homogeneous equation is 

$$y'' + p(x)y' + q(x)y = 0$$

It is also called the reduced equation. Its solution is $y_c = c_1y_1 + c_2y_2$ and is part of inhomegeneous equation solution. The name of the solution is complementary solution. Some called solution to the associated homogeneous equation, $y_h$ 


## Example

Spring-mass-dashpot

$$mx'' + bx' + kx = f(t)$$

$$mx'' = -kx -bx' + f(t)$$

$f(t)$ is external force. When external force is zero, the system just respond to the initial condition and does what it wants. The other model is we don't let it respond what it wants, instead, force it from the outside by pushing with external force. 

Forced system when $f(t) \neq 0$, passive system when $f(t) = 0$


## Example

A RCL circuit. 

Sum of voltage drops is zero if no external power source

Voltage drop on inductor + voltage drop on resister + voltage drop on capcitor equals the voltage source.

$$Li' + Ri + \frac{q}{C} = E(t)$$

If $E(t)$ is little dry cell, it is a passive circuit, $E'(t) = 0$

$$Li'' + Ri' + \frac{i}{C} = E'(t)$$



## Theorem of $Ly = f(x)$

$L$ is linear operator. The solution has this form $y_p + y_c$, so $y = y_p + c_1y_1 + c_2y_2$

$y_p$ is the particular solution to $Ly = f(x)$

To solve the equation, first find complementary solution, then find particular solution

To prove: 

1. All the $y_p + c_1y_1 + c_2y_2$ are solutions

$$L(y_p + c_1y_1 + c_2y_2) = L(y_p) + L(c_1y_1 + c_2y_2) = f(x)$$

$L(c_1y_1 + c_2y_2) = 0$ because it is a solution, $ L(y_p) = f(x)$ because it is a particular solution. Therefore it satisfies, $L(y) = f(x)$


2. No other solutions

Suppose $u(x)$ is a new solution, then $L(u) = f(x)$, we know what $L(y_p) = f(x)$, then

$$L(u - y_p) = L(u) - L(y_p) = f(x) - f(x) = 0$$

Therefore $u - y_p$ is a solution of homonegeous equation and equals $k_1y_1 + k_2y_2 \implies u = y_p + k_1y_1 + k_2y_2$

This means $u(x)$ is not new, contradicts with the assumption


## Back to First-order Equation

$$y' + ky = q(t)$$

Solution look like $y = e^{-kt}\int q(t)e^{kt}\ dt + ce^{-kt}$

Is $ce^{-kt}$ a complementary solution?. Yes, because solution of $y' + ky = 0$ is $ce^{-kt}$

Is $e^{-kt}\int q(t)e^{kt}\ dt$ a particular solution? Yes, make it particular by changing the integral to definite integral on a particular range

For $k > 0$, y = steady state + transient. $t \to \infty, ce^{-kt} \to 0$

For $k < 0$, this analysis does not work


## Second-order Equation

$$y'' + Ay' + By = f(t)$$

A, B are constant

Solution $y = y_p + c_1y_1 + c_2y_2$

Complementary solution contains the initial conditions. When does $c_1y_1 + c_2y_2 \to 0 $ as $t \to \infty$ for all $c_1, c_2$?

If this is solved, ODE is called stable. We know how it behaves it the long term.


## Characteristic roots, Solution, Stability condition

- Roots: $r_1 \neq r_2$

Solution: $c_1 e^{r_1t} + c_2 e^{r_2t}$

Stability condition: $c_1 < 0, c_2 < 0$


- $r_1 = r_2$

$(c_1 + c_2t)e^{rt}$

$r_1 < 0$


- $r = a \pm bi$

$e^{at}(c_1\cos bt + c_2\sin bt)$

$a < 0$

Conclusion: ODE is stable if all the characteristic roots have negative real parts
