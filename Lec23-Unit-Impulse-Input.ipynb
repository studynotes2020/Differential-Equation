# Unit Impulse Input

Impulse of $f(t)$ over $[a, b]$ is $$\int_a^b f(t)\ dt$$ 

If $f(t)$ is constant F, then impulse is $F.(b - a)$

When we say apply an impulse, it means apply constant force over a definite interval. 

An impulse is the unit area under the curve F vs t. So if the length is h, then the height is 1/h. Area = h * 1/h = 1. In other words, the narrower the t, the higher the f


## Typical Spring Problem

$$y'' + y = \frac{1}{h}u_{oh}(t)$$

$u_{oh}(t)$ is unit box function from 0 to h

What is LP of the above RHS?

$$L[\frac{1}{h}[u(t) - u(t - h)]] = \frac{1}{h}[\frac{1}{s} - \frac{e^{-hs}}{s}] = \frac{1 - e^{-hs}}{hs}$$

Let h goes to zero.

$$\lim \limits_{h \to 0}\frac{1 - e^{-hs}}{hs} = \lim \limits_{h \to 0}\frac{e^{-hs}}{1} = 1 $$

When $h \to 0$, we define $$\frac{1}{h}u_{oh}(t) \to \delta(t)$$ 

called Dirac delta, which has Laplace transform, 

$$L(\delta(t)) = 1$$

It is the same to integrate from zero to infinity. In order to avoid controversy, make it from negative infinity to infinity

$$\int_{-\infty}^\infty \delta(t)\ dt = 1$$

What is its convolution, $f(t) * \delta(t)$ ?

$$L[f(t) * \delta(t)] = F(s).1$$

To avoid ambiguity of $f(t)$ when it is negative, we make it unique

$$L[u(t)f(t) * \delta(t)] = F(s).1$$

So $\delta(t)$ for convolution operation is acting like an identity

Differentiate unit step function is delta function

$$u'(t) = \delta(t)$$


## Back to ODE

$$y'' + y = A\delta(t - \frac{\pi}{2})$$

Initial condition $y(0) = 1, y'(0) = 0$.

The equation means the spring is kicked with impulse A at $t = \pi /2$ 

Take LP of both sides

$$s^2Y - s + Y = Ae^{-\pi s /2}$$

$$Y = \frac{s}{s^2 + 1} + \frac{Ae^{-\pi s/2}}{s^2 + 1}$$

Inverse LP

$$y = \cos t + u(t - \pi/2)A\sin(t - \pi/2)$$

$$y = \begin{cases}
\cos t, 0 \leq t \leq \pi/2 \\
(1 - A)\cos t, t \geq \pi/2
\end{cases}$$

```julia
using Plots
f(t) = cos(t)
g(t) = -cos(t)  # A = 2
plot(f, 0, pi/2)
plot!(g, pi/2, 2pi)

g2(t) = -3cos(t) # A >> 1
g3(t) = 2cos(t) # A < 1
g4(t) = 0  # A = 1

plot!(g2, pi/2, 2pi)  
plot!(g3, pi/2, 2pi)  
plot!(g4, pi/2, 2pi)  
```

$$y'' + ay' + by = f(t)$$

$f(t)$ is the input like sin, cosine, exponential, or sum of those

Initial condition $y(0) = 0, y'(0) = 0$. Taking LP

$$s^2Y + asY + bY = F(s)$$

$$Y = F(s).\frac{1}{s^2 + as + b}$$

Inverse LP

$$y(t) = f(t) * w(t)$$

Define transfer function

$$W(s) = \frac{1}{s^2 + as + b}$$

$$L^{-1}(W(s)) = L^{-1}(\frac{1}{s^2 + as + b}) = w(t)$$

weight function of the system. Therefore

$$y(t) = f(t) * w(t) = \int_0^t f(u)w(t - u)\ du$$

So, the solution of the equation is presented as definite integral. Then the next question is what is w(t)?

Let starts with modeling the system with a kick at time zero with same initial condition

$$y'' + ay' + by = \delta(t)$$

Taking Laplace transform and we will get the result

$$Y = \frac{1}{s^2 + as + b}$$

So, the inverse LP is

$$y(t) = L^{-1}[\frac{1}{s^2 + as + b}] = w(t)$$

The weight function is the response of the system at rest to a sharp kick at time zero with unit impulse. Convolution of f(t) with w(t) can be thought of as the superposition of a lot of sharp kick continuously
