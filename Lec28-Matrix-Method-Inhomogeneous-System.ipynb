# Matrix Method to Solve Inhomogeneous System

Theory: $$\vec x' = A\vec x$$

2x2 matrix and matrix A is constant

General solution to $\vec x' = A\vec x$ is $\vec x = c_1\vec x_1 + c_2\vec x_2$ where $\vec x_1, \vec x_2$ are two linearly independent solutions

Both vector x1, x2 are function of t

Easy to show: All these (linearity + superposition) are solutions

Hard to show: There are no other solutions

Wronskian of two solutions, 

$$W(\vec x_1, \vec x_2) := \begin{vmatrix}\vec x_1 & \vec x_2 \end{vmatrix}$$

Theorem: Either W(t) is zero for all t, if x1 and x2 are dependent. Or never zero for any t, if x1 and x2 are independent


## Fundamental Matrix for $\vec x' = A\vec x$

Two columns are two solutions.

$$X := \begin{bmatrix} \vec x_1 & \vec x_2 \end{bmatrix}$$

$\vec x_1, \vec x_2$ are independent solutions

Properties:

1. determinant $|X| \neq 0$ for any t
2. $X' = AX \iff \begin{bmatrix} \vec x_1' & \vec x_2' \end{bmatrix} = A\begin{bmatrix} \vec x_1 & \vec x_2 \end{bmatrix} = \begin{bmatrix} A\vec x_1 & A\vec x_2 \end{bmatrix}$

$\iff \vec x_1' = A\vec x_1, \vec x_2' = A\vec x_2$

Take two equations make them into one solution, by putting them into a matrix



## Inhomogeneous System

$$x' = ax + by + r_1(t)$$
$$y' = cx + dy + r_2(t)$$

Inhomogeneous means there are extra r(t) term on RHS. The matrix abbreviation will look like

$$\vec x' = A\vec x + \vec r(t)$$

Theorem: $\vec x_{gen} = \vec x_c + \vec x_p$. The general solution is equal to sum of complementary solution (solution to homogeneous system x' = Ax) and particular solution. Linearity + superposition

We have done the work to look for $\vec x_c$, so now the work is to find $\vec x_p$


Consider the equation
$$x'' + x = \tan t$$

Mixing problem is an example of inhomogeneous system.

Flow rate: liters/hour. 

x is amount of salt in the tank 1. y is amount of salt in the tank 2.

Let 1 and 2 flow in tank 1, then 3 flow out in tank 1. 1 and 3 flow into tank 2, then 2 and 2 flow out tank 2.

Let say salt concentration of $5e^{-t}$ flow in on one end (steadily decline concentration) and salt concentration 0 (pure water) flow in on another end

Build the model
$$x' = -3x + 2y + 5e^{-t}$$
$$y' = 3x - 4y$$

$$\vec x' = \begin{bmatrix}-3 & 2 \\ 3 & -4 \end{bmatrix} \vec x + 
\begin{bmatrix}5e^{-t} \\ 0 \end{bmatrix}$$

## Method to Solve $\vec x' = A\vec x + \vec r$, find $\vec x_p$

The method is called variation of parameters. $c_1, c_2$ becomes function of t (varying)

$$\vec x_p = v_1(t)\vec x_1 + v_2(t)\vec x_2$$

$$\vec x_p = X\vec v = \begin{bmatrix}x_1 & x_2 \\ y_1 & y_2 \end{bmatrix} \begin{bmatrix}v_1 \\ v_2 \end{bmatrix}  $$

Substitute x' into the system and see what v is. Because $\vec x_p' = A\vec x_p + \vec r$ and $\vec x_p' = (X\vec v)'$

$$X'\vec v + X\vec v' = AX\vec v + \vec r$$

From the homonegeneous solution we know that $X'\vec v = AX\vec v$, hence 

$$X\vec v' = \vec r$$

$$\vec v' = X^{-1}\vec r$$

For X to have inverse, the determinant in not zero. Because the column vectors are independent solutions.

$$\vec v = \int X^{-1} \vec r\ dt$$

Integrate each entry in X inverse

Particular solution: $$\vec x_p = X\vec v = X\int X^{-1} \vec r\ dt$$

one particular solution is good enough, so no need to put in constant after integration
