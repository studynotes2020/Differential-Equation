$$\begin{bmatrix} x \\ y \end{bmatrix}' =

Starts with example

$$x' = -x + by$$
$$y' = cx - 3y$$

Let say first equation represents Massachusett. Second equation represents New Hampshire.

x, y represent departures from normal tourise advertising budget. If both zeroes, that means both spend normally. If x is positive, it means Massachusett decided to spend more. Negative coefficient -1 and -3 is like a pull back to normal. If the city council spend more, there is a certain poll yelling to spend less. -3 means city in New Hamsphire is more sensitive to change of budget. Coefficient b, c is like responding to other city's budget. If b is small, Massachusett follows a little of budget New Hamsphire.

Another example

$$x' = -x + 2y$$
$$y' =  - 3y$$

Zero coefficient means New Hampshire doesn't care about budget of Massachusett. What is the result of this model?

$$A = \begin{bmatrix} -1 & 2 \\ 0 & 3 \end{bmatrix}$$

$$\lambda^2 + 4\lambda + 3 = 0$$

$\lambda = -3, -1$ eigenvalues

$\lambda = -3 \implies 2a_1 + 2a_2 = 0 \implies \vec a = \begin{bmatrix}1\\ -1 \end{bmatrix}$

$\lambda = -1 \implies 0a_1 + 2a_2 = 0 \implies \vec a = \begin{bmatrix}1 \\ 0 \end{bmatrix}$

General Solutions:

$$\begin{bmatrix} x \\ y \end{bmatrix} =
c_1\begin{bmatrix} 1 \\ -1 \end{bmatrix} e^{-3t} + 
c_2\begin{bmatrix} 1 \\ 0 \end{bmatrix} e^{-t} $$

Expand it

$$x = c_1e^{-3t} + c_2e^{-t}$$
$$y = -c_1e^{-3t}$$

Then plot the vector

1. 4 easy solutions. Ex. $c_1 = \pm 1, c_2 = 0$ and $c_1 = 0, c_2 = \pm 1$
2. Fill in the rest

As t goes to infinity, the solutions go to zero. But because first term is $e^{-3t}$, it goes down more quickly, so second is the dominant term when t goes to infinity. But when t goes to negative infinity, the first term is the dominant term

All the trajectory will be end up in the origin, so called nodal sink or sink node. This is called stable. There is also source node

Let say if A produces solution

$$\vec x = c_1\begin{bmatrix} 1 \\ -1 \end{bmatrix} e^{3t} + 
c_2\begin{bmatrix} 1 \\ 0 \end{bmatrix} e^{t} $$

Then all arrows will be reversed. Origin becomes source node. The solution is unstable

### Another Example

$$x' = -x + 3y$$
$$y' = 5x - 3y$$

$$A = \begin{bmatrix} -1 & 3 \\ 5 & -3 \end{bmatrix}$$

$$\lambda^2 + 4\lambda - 12 = 0$$

$$(\lambda + 6) (\lambda - 2) = 0$$

$\lambda = -6, 2$

$\lambda = -6 \implies 5a_1 + 3a_2 = 0 \implies \vec a = \begin{bmatrix}3\\ -5 \end{bmatrix}$

$\lambda = 2 \implies -3a_1 + 3a_2 = 0 \implies \vec a = \begin{bmatrix}1\\ 1 \end{bmatrix}$

General Solutions:

$$\begin{bmatrix} x \\ y \end{bmatrix} =
c_1\begin{bmatrix} 3 \\ -5 \end{bmatrix} e^{-6t} + 
c_2\begin{bmatrix} 1 \\ 1 \end{bmatrix} e^{2t} $$

The curve is called saddle. It is unstable.

In this model, two cities x and y, both want to outspend others, indicated by coefficients +3 and +5. They both lose in the end, because they end up all spending an infinity of dollars and nobody gets any more tourists than anybody else

### Another Example

$$x' = -x - y$$
$$y' = 2x - 3y$$

-y here means when y increasing budget, x lower the budget

$$A = \begin{bmatrix} -1 & -1 \\ 2 & -3 \end{bmatrix}$$

$$\lambda^2 + 4\lambda + 5 = 0$$

$\lambda = -2 \pm i$

So the solution has $e^{-2t}.e^{it}$

$$\begin{bmatrix} x \\ y \end{bmatrix} = 
c_1\left(\begin{bmatrix} a_1 \\ a_2 \end{bmatrix} \cos t +
\begin{bmatrix} b_1 \\ b_2 \end{bmatrix} \sin t \right) e^{-2t} +
c_2(...e^{-2t}.e^{-it})$$

If plot vector curve, because inside () there are cosine and sine, so it is bounded between 0 and 1, with period $2\pi$. It satisfies equation $Ax^2 + Bx^2 + Cxy = D$. To figure out A, B, C, D, square the x and collect the term with equal polynomials.

The curves are ellipses and spiral towards origin. Spiral sink.

(Example of Conic setion includes hyperbola, parabola, ellipses)

To know the spiral spins clockwise or anti-clockwise, put x = 1, y = 0, in the equation then the vector is $\begin{bmatrix} -1 \\ 2 \end{bmatrix}$ which indicates anti-clockwise direction
