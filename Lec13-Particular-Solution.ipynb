# Finding Solution for Inhomogeneous ODE

$$y'' + Ay' + By = f(x)$$

Find a particular solution $y_p$. Then the general solution is $y = y_p + c_1y_1 + c_2y_2$

Important $f(x):$ $e^{ax} (a < 0)$, $\sin \omega x, \cos\omega x$, $e^{ax} \sin \omega x, e^{ax} \cos \omega x$

They are all special cases of $e^{(a + ib)x}$

$e^{(a + ib)x}$ can be simplied as $e^{\alpha x}$ with $\alpha$ complex number and this is going to be our input

Think of the left hand side as polynomial operator

$$(D^2 + AD + B)y = f(x)$$

with $p(D) = D^2 + AD + B$

An important equation that we will be using. Substitution rule

$$p(D) e^{\alpha x} = p(\alpha) e^{\alpha x}$$

Prove: 
$$\begin{align}(D^2 + AD + B)e^{\alpha x} & = D^2 e^{\alpha x} + ADe^{\alpha x} + Be^{\alpha x} \\
& = \alpha^2 e^{\alpha x} + A\alpha e^{\alpha x} + Be^{\alpha x} \\
& = e^{\alpha x}(\alpha^2 + A\alpha + B) \\
& = p(\alpha) e^{\alpha x}
\end{align}$$

## Exponential Input Theorem

$$y_p = \frac{e^{\alpha x}}{p(\alpha)}$$

Prove if I apply $p(D)$ to $y_p$, do I get $e^{\alpha x}$: 

$$p(D)y_p = e^{\alpha x}$$

$$p(D) \frac{e^{\alpha x}}{p(\alpha)} = \frac{p(\alpha)e^{\alpha x}}{p(\alpha)} = e^{\alpha x}$$

If $p(\alpha) = 0$, this won't work. We will cater it later, for the time being assume $p(\alpha) \neq 0$


### Example

Find the particular solution for $$y'' - y' + 2y = 10e^{-x}\sin x$$

Start with polynomial operator and complexify the input. $e^{-x}\sin x$ is the imaginary part of $e^{(-1 + i)x}$

$$(D^2 - D + 2)\tilde y = 10e^{(-1 + i)x}$$

$\tilde y$ to indicate the solution we get is not the solution for original equation. From the theorem, we get

$$\tilde y_p = \frac{10e^{(-1 + i)x}}{(-1 + i)^2 - (-1 + i) + 2}$$

$$\tilde y_p = \frac{10e^{(-1 + i)x}}{3 - 3i} = \frac{10}{3}\frac{i + 1}{2}e^{-x}(\cos x + i\sin x)$$

$$\begin{align}
y_p & = Im(\tilde y_p) \\
& = \frac53 e^{-x} (\cos x + \sin x) \\
& = \frac53 e^{-x} \sqrt{2} \cos(x - \frac{\pi}{4})
\end{align}$$


### Case $p(a) = 0$

Change variable $\alpha$ to $a$ for easier writing.

Exponential-shift rule $$p(D)e^{ax} u(x) = e^{ax} p(D + a) u(x)$$

This rule is just direct consequence of differentiation product rule

To prove the rule, just starts with simple $p(D) = D$

$$\begin{align}
De^{ax}u & = e^{ax}Du + ae^{ax}u \\
& = e^{ax}(D + a)u
\end{align}$$

If this is true also for $D^2$?

$$\begin{align}
D^2e^{ax}u & = D(De^{ax}u) \\
& = D(e^{ax}(D + a)u \\
& = e^{ax}(D + a)(D + a)u \\
& = e^{ax}(D + a)^2 u
\end{align}$$

We can see that the process can be generalized to n-th derivative by induction



Back to the form

$$(D^2 + AD + B)y = e^{ax}$$

with $a$ is allowed to be complex and $p(x) = 0$. 

$$y_p = \frac{xe^{ax}}{p'(a)}$$

$a$ is a simple root of $p(D)$. Is $p'(a)$ guaranteed no zero? Yes

Proof: if $a$ is simple root case, it can be factored out, so

$$p(D) = (D - b)(D - a), b \neq a$$

By product rule

$$p'(D) = (D - a) + (D - b)$$

$$p'(a) = a - b$$

This proves that if $a$ is simple, $p'(a)$ is guaranteed not to be zero

What if $a$ is double root?

In that case, $$y_p = \frac{x^2 e^{ax}}{p''(a)}$$

Things go on for higher root


### Now propose $y_p = \frac{e^{ax}x}{p'(a)}$ is the solution

We need to check if the following statement is true

$$p(D)y_p = p(D) \frac{e^{ax}x}{p'(a)} = e^{ax}$$

By exponential-shift rule

$$e^{ax}\frac{(D + a - b)Dx}{p'(a)} = e^{ax}\frac{a - b}{a - b} = e^{ax}$$

### Example

$$y'' - 3y' + 2y = e^x$$

$a = 1$ is simple root of $D^2 - 3D + 2$. If this is double root, ODE won't look like that

$$p'(D) = 2D - 3 \implies p'(1) = -1$$

$$y_p = \frac{xe^x}{-1} = -xe^x$$
