# Fourier analysis

We are able to break down a messy signal into pure oscillations

$f(t)$ is sum of sines of various frequency $b_1\sin w_1t + b_2\sin w_2t + b_3\sin w_3t + ...$

For a square wave (amplitude from 0 to 1, half period is 1), what is its fourier series?

It is neither even nor odd. So, try to reduce the amplitude (shift down) by half, then it becomes an odd function

We have a function $g(u)$ calculated in first lecture of fourier series, signal goes from -1 to 1, square wave with half period pi

$$g(u) = \frac{4}{\pi} \sum_{n=odd}^{\infty} \frac{\sin nu}{n}$$


```julia
using Plots
g(x) = (4/pi)*(sin(x)  + sin(3x)/3  + sin(5x)/5)

plot(g, -pi, 2pi)
```

This curve looks like the curve of square wave half period 1. So how do I convert this known solution for the first problem

Change the variable, $u = \pi t$. When $t = 1, u = \pi$

$$f(t) = S(t) + \frac12 $$

$$S(t) = \frac12 g(u)$$

$$f(t) = \frac12 + \frac{2}{\pi} \sum_{odd} \frac{\sin n\pi t}{n}$$




```julia
f(t) = 1/2 + 2/pi *(sin(pi*t) + sin(3pi*t)/3 + sin(5pi*t)/5)
plot(f, -pi, 2pi)
```

Solve this to find a particular solution $x_p$

$$x'' + w_0^2 x' = f(t)$$

We can find $x_p$ of RHS if either cosines or sines or multiples of them. The procedure is to use the complex exponential $e^{iwt}$

The solution is $$x_p = \frac{\text{combination of cos wt, sin wt}}{w_0^2 - w^2}$$

If $f(t) = \frac{a_0}{2} + \sum_1^{\infty} a_n\cos w_nt + b_n \sin w_nt$, where the general case $w_n = \frac{n\pi}{L}$ period is 2L

then $$x_p = \frac{a_0}{2w_0^2} + \sum_1^{\infty} \frac{a_n\cos w_nt + b_n\sin w_nt}{w_0^2 - w_n^2}$$ 

In other words, when we have fourier series for the input, we immediately have the fourier series of the response

Input: $f(t)$ is a square wave from 0 to 1, half period 1 and the equation is $x'' + w_0^2x = f(t)$

Response: $$x_p = \frac{1}{2w_0^2} + \frac{2}{\pi}\sum_{odd} \frac{\sin n\pi t}{n(w_0^2 - (n\pi)^2)}$$ 

Suppose natural frequency $w_0 = 9$, $x_p(t) \approx 0.00617 + 0.6366(sin(pi*t)/71  + sin(3pi*t)/-23 + sin(5pi*t)/-828)$

$x_p(t) \approx 0.00617 +  ...$

If $w_0 = 9$, $n = 3$ is the resonant frequency

If $w_0 = 10$, near resonance occurs for frequency $3\pi$ in the input

Frequency in $f(t)$ are hidden but the system picks out (resonate with) the frequency closest to its natural frequency

## Book's method (notes also)

Use differentiation of Fourier series term by term. Assume solution of the form $x_p = c_0 + \sum_1^{\infty} c_n \sin n\pi t$. The only question is what are the $c_n$? 

The general method is substitute into the ODE

$x_p'' = \sum_1^{\infty} -c_n(n\pi)^2\sin n\pi t$

multiply $w_0$ to $x_p$ then sum them up

$$x_p'' + w_0x_p = \text{Fourier Series for $f(t)$}$$

By equating coefficients, get $c_n$

$$w_0^2c_0 + \sum_1^{\infty} c_n(w_0^2 - (n\pi)^2)\sin n\pi t = \frac12 + \frac{2}{\pi} \sum_{odd} \frac{\sin n\pi t}{n}$$

$$c_0 = \frac{1}{2w_0^2}$$
