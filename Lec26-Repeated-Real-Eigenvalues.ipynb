# A Fish Tank

Suppose there is a fish tank with three sections

$x_i =$ temperature in tank i, find three functions

$$x_1' = a(x_3 - x_1) + a(x_2 - x_1)$$
$$x_1' = -2ax_1 + ax_2 + ax_3$$

take a = 1 and build the model for other two regions

$$x_1' = -2x_1 + x_2 + x_3$$
$$x_2' = x_1 - 2x_2 + x_3$$
$$x_3' = x_1 + x_2 - 2x_3$$

$$|A - \lambda I| = \begin{vmatrix} -2-\lambda & 1 & 1 \\ 1 & -2-\lambda & 1 \\ 1 & 1 & -2-\lambda \end{vmatrix} = 0$$

$$-(\lambda + 2)^3 + 2 - 3(-2 - \lambda) = 0$$

$$\lambda^3 + 6\lambda^2 + 9\lambda = 0$$

$$\lambda(\lambda + 3)^2 = 0$$

$\lambda_1 = 0, \lambda_2 = -3$ (double roots)

When $\lambda = 0$, find its associate eigenvector

$$-2a_1 + a_2 + a_3 = 0$$
$$a_1 - 2a_2 + a_3 = 0$$
$$a_1 + a_2 - 2a_3 = 0$$

Solution is $\begin{bmatrix}a_1 \\ a_2 \\ a_3 \end{bmatrix} e^{\lambda t}$

Because $\lambda = 0, e^{\lambda t} = 1$, the solution is a constant solution. All temperature is the same, $a_1 = a_2 = a_3 = 1$ or $a_1 = a_2 = a_3 = 20$ can be the solution

When $\lambda = -3$, 

$$a_1 + a_2 + a_3 = 0$$
$$a_1 + a_2 + a_3 = 0$$
$$a_1 + a_2 + a_3 = 0$$

There are many solutions, suppose we take $\vec a = \begin{bmatrix}1 \\ 0 \\ -1 \end{bmatrix}$, then the solution is $\begin{bmatrix}1 \\ 0 \\ -1 \end{bmatrix} e^{-3t}$. Another solution is $\begin{bmatrix}1 \\ -1 \\ 0 \end{bmatrix} e^{-3t}$. Then we have two independent solution just by picking two different vectors

General Solution:

$$\begin{bmatrix}x_1 \\ x_2 \\ x_3 \end{bmatrix} = 
c_1 \begin{bmatrix} 1\\ 0 \\ -1 \end{bmatrix}e^{-3t} + 
c_2 \begin{bmatrix}1 \\ -1 \\ 0 \end{bmatrix} e^{-3t} + 
c_3 \begin{bmatrix}1 \\ 1 \\ 1 \end{bmatrix}$$

As $t \to \infty$, first two term go to zero, last term remain which mean all temperature is the same

Though $\lambda$ is repeated eigenvalue, can find enough independent eigenvalues to make up needed number of independent vectors. This is called complete eigenvalues. Defective is when you cannot find enough independent eigenvectors. Then we are in trouble, we cannot find enough solutions.


Theorem: A real nxn matrix which is symmetric ($A^T = A$) $\implies$ all its eigenvalues are complete


## Complex Eigenvalues

- Calculate complex eigenvectors
- Form solutions $\vec a e^{(a + bi)t}$ where $\vec a$ is complex
- Take real and imaginary part

In this example, eigenvalues are pure imaginary

$$x' = x + 2y$$
$$y' = -x - y$$

$$A = \begin{bmatrix}1 & 2 \\ -1 & -1 \end{bmatrix}$$

Characteristic equation

$$\lambda^2 + 1 = 0$$

$$\lambda = \pm i$$

Find eigenvectors

$$(1 - i)a_1 + 2a_2 = 0$$
$$-a_1 + (-1-i)a_2 = 0$$

Take $a_1 = 1$ then $a_2 = \frac{i-1}{2}$

Complex Solution: 
$$\begin{bmatrix}1 \\ \frac{i-1}{2} \end{bmatrix} e^{it} = 
(\begin{bmatrix}1 \\ \frac{-1}{2} \end{bmatrix} + i \begin{bmatrix}0 \\ \frac12 \end{bmatrix}) (\cos t + i\sin t)$$

Real part: $$\begin{bmatrix}x \\ y \end{bmatrix} = 
\begin{bmatrix}1 \\ -\frac{1}{2} \end{bmatrix} \cos t - \begin{bmatrix}0 \\ \frac{1}{2} \end{bmatrix} \sin t$$

$$x = \cos t, y = -\frac12 (\cos t + \sin t)$$

When plot y vs x, it is an ellipse curve
