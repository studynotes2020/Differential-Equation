# Linear Second Order ODE with Constant Coefficients

$$y'' + Ay' + By = 0$$

Second order because y differentiated two times. A and B is constant coefficients. RHS zero makes it homogeneous. Study this first, then we can go to inhomogeneous equation

Assume general solution

$$ y = c_1y_1 + c_2y_2$$

Because this is second order, the solution will be integrated twice, so two constant. Number of constant equals to the order of the equation.

$y_1, y_2$ is the solution

Initial condition are satisfied by choosing $c_1, c_2$


## Spring-Mass-Dashport System

Equation of the system: Force = spring + dashpot

Force is proportional to x varies. Velocity of dashpot resist the force. 

$$mx'' = -kx - cx'$$

$$mx'' + cx' + kx = 0$$

Put it in standard form.

$$x'' + \frac{c}{m}x' + \frac{k}{m}x = 0$$

$k/m$ is spring constant, $c/m$ is damping constant

### To solve ODE, find two solutions

$y_1, y_2$ should be independent (not multiple of each other)

Basic method: try $y = e^{rt}$, t is independent

Plug in $$r^2e^{rt} + Are^{rt} + Be^{rt} = 0$$
 
$e^r$ is always positive, regardless of t, so we can cancel it out

$$r^2 + Ar + B = 0$$

This is called the characteristic equation of the system.

There are several cases and each case correspond to a different behavior. The possibilities of roots are

1. Real and distinct
2. A pair of complex conjugates
3. Real and equal

## Case 1: Real and Distinct

Solution: $$y = c_1e^{r_1t} + c_2e^{r_2t}$$

Example: $$y'' + 4y' + 3y = 0$$

Initial condition $y(0) = 1$ starts at distance one unit to the right, $y'(0) = 0$ starts at rest

Characteristic equation: $$r^2 + 4r + 3 = 0$$

$$(r + 3)(r + 1) = 0$$

Solution: $$y = c_1e^{-3t} + c_2e^{-t}$$

There are two initial conditions, so we can get values for two constants.

From the first condition, substitute: $$1 = c_1 + c_2 \ --- (1)$$

Differentiate the solution: $$y' = -3c_1e^{-3.t} - c_2e^{t}$$ 

Then substitute in the second condition $$0 = -3c_1 - c_2 \ --- (2)$$

Solve (1) and (2), we get $$c_1 = -\frac{1}{2}, c_2 = \frac32$$

Solution: $$y = -\frac{1}{2}e^{-3t} + \frac32 e^{-t}$$

```julia
using Plots
y(t) = -1/2*exp(-3t) + 3/2*exp(-t)  # overdamped
plot(y, 0, 2pi)
# if apply strong negative push, curse go under x-axis and damped 
# if apply positive push, starts with positive slope then damped
```

## Case 2: Complex Roots

$$r = a \pm bi$$

We get complex solution $$y = e^{(a + bi)t}$$

Theorem: if $u + iv$ is complex solution to $y'' + Ay' + By = 0$, then $u, v$ are real solutions

Proof: $$(u + iv)'' + A(u + iv)' + B(u + iv) = 0$$

$$u'' + Au' + Bu + i(v'' + Av' + Bv) = 0$$

The equation says that real part plus imaginary part is zero. That happen when real part is zero and imaginary part is zero. When real part is zero, u is the solution. Similarly when imaginary part is zero, v is the solution.

$$y = e^{at + ibt}$$

Real part: $e^{at}\cos bt$. Imaginary part: $e^{at}\sin bt$

Solution: $$y = e^{at}(c_1\cos bt + c_2\sin bt)$$

Notice that there are cos and sin with same frequency b, so this is an oscillation with frequency b, though the amplitude is different.

Example:

$$y'' + 4y' + 5y = 0$$
Initial condition: $y(0) = 1, y'(0) = 1$

$$r^2 + 4r + 5 = 0$$
$$r = \frac{-4 \pm \sqrt{-4}}{2} = -2 \pm i$$

Solution: $$y = e^{-2t}(c_1\cos t + c_2\sin t)$$

Using the initial conditions to solve for constant,

Solution: $$y = e^{-2t}(\cos t + 2\sin t) = \sqrt{5}e^{-2t}\cos(t-\phi)$$

Using trig identities, $\cos(x - y) = \cos x \cos y + \sin x \sin y$ .We get $\sqrt{5}$ and $\phi$ roughly 63 degree

```julia
a(t) = sqrt(5)*exp(-2*t)
plot(a, 0, pi)
b(t) = cos(t-1.10714871)  
plot!(b, 0, 2pi)
# not able to see the oscillation bcs frequency lower than damping constant
```

```julia
plot!(t->a(t)*b(t), 0, 2pi)  
# underdampe as cos oscillate. Not able to see from graph
```

## Case 3: Equal Roots

Critically damped.

$r^2 + Ar + B = 0$ has to equal roots, $r = -a$

$$(r + a)^2 = 0$$
$$r^2 + 2ar + a^2 = 0$$

General solution for ODE of this form: $$y'' + 2ay' + a^2y = 0$$

In other words, the damping and spring constant is related in a special way

Solution: $$y_1 = e^{-at}$$

There are four ways. Here introduce one way. Know one solution ($y_1$) to $y'' + py' + qy = 0$. Another solution of the form $y_2 = y_1u$

$$y_2 = e^{-at}u$$
$$y' = -ae^{-at}u + e^{-at}u'$$
$$y'' = a^2e^{-at}u - 2ae^{-at}u' + e^{-at}u''$$

Multiply first line with $a^2$, second line with $2a$ and add them up

$$y'' + 2ay' + a^2y = a^2e^{-at}u - 2ae^{-at}u' + e^{-at}u'' + 2a(-ae^{-at}u + e^{-at}u') + a^2(e^{-at}u) = 0$$

$$\begin{align}e^{-at}u'' & = 0 \\
u'' & = 0 \\
u & = c_1t + c_2 
\end{align}$$

This gives me whole family of solution, but just t is enough. Because we only need one solution different from $e^{-at}$, so

$$y = e^{-at}t$$
