## Theory of Second Order Linear Homogenerous ODE

$$y'' + p(x)y' + q(x)y = 0$$

linear in $y, y', y''$

Solution method was to find $y_1, y_2$ which are independent solutions

Independent means $y_2 \neq cy_1, y_1 \neq c'y_2$. Say this in two statement because I want to exclude $y_1 = 0, y_2 \neq 0$

Then all solution are $y = c_1y_1 + c_2y_2$. Why?

Question 1: Why are $c_1y_1 + c_2y_2$ solutions?

Question 2: Why all the solutions? (why is there no other solutions)

Answer 1: Superposition


## Superposition principle

$y_1, y_2$ are solutions to linear homogeneous ODE that implies $c_1y_1 + c_2y_2$ (their linear combination) is a solution

Proof: 
$$y'' + py' + qy = 0$$

$$D^2y + p Dy + qy = 0$$

$$(D^2 + pD + q)y = 0$$

Define linear operator $L = D^2 + pD + q$

$$Ly = 0$$

Think L as a black box. Input is function of x $u(x)$, output is another function of x $v(x)$. Solve differential equation means what $u(x)$ to put in so that $v(x) = 0$

Linear operator means $L(u_1 + u_2) = L(u_1) + L(u_2)$ and $L(cu) = cL(u)$

$D$ is linear operator because $(u_1 + u_2)' = u_1' + u_2'$ and $(cu)' = cu'$


### Proof of superposition

ODE: $Ly = 0$

$L(c_1y_1 + c_2y_2) = L(c_1y_1) + L(c_2y_2) = c_1L(y_1) + c_2L(y_2)$

$y_1, y_2$ are solutions implies $L(y_1) = 0, L(y_2) = 0$. So the sum of zero is zero

This proves $L$ is linear



## Solve Initial Value Problem

$c_1y_1 + c_2y_2$ is enough to satisfy any initial values

Proof:

Inital values are 

$y(x_0) = a$

$y'(x_0) = b$

$y = c_1y_1 + c_2y_2$

$y' = c_1y_1' + c_2y_2'$

Plug in $x = x_0$

$c_1y_1(x_0) + c_2y_2(x_0) = a$

$c_1y_1'(x_0) + c_2y_2'(x_0) = b$

$c_1, c_2$ are variables to be found. The simultaneous equation is solvable if determinant evaluated at $x_0$ is not zero

$$\begin{vmatrix} y_1 & y_2 \\ y_1' & y_2' \end{vmatrix}_{x_0} \neq 0$$

This determinant is called Wronskian $W(y_1, y_2)$ where $y_1, y_2$ are function of $x$

If $y_2 = cy_1$ (dependent), Wronskina is zero for all $x$

$$\begin{vmatrix} y_1 & y_2 \\ y_1' & y_2' \end{vmatrix} = \begin{vmatrix} y_1 & cy_1 \\ y_1' & cy_1' \end{vmatrix} = y_1cy_1' - cy_1y_1' = 0$$

Theorem: if $y_1, y_2$ are solutions to ODE, either $W(y_1, y_2) = 0$ for all x, or $W(y_1, y_2)$ is never zero for all x

$c_1y_1 + c_2y_2$ is all the solutions = $c_1'u_1 + c_2'u_2$ where $u_1, u_2$ are any other pair of independent solutions which can be written in terms of $y_1, y_2$

$u_1 = k_1y_1 + k_2 y_2$

$u_2 = k_3y_1 + k_4y_2$

$y_1, y_2$ are typical solution that we get like exponential, sine, cosine.

Finding normalized solutions at 0, $Y_1, Y_2$. They are solutions satisfy special initial condition

$Y_1: Y_1(0) = 1, Y_1'(0) = 0$

$Y_2: Y_2(0) = 0, Y_1'(0) = 1$

Suppose equation is

$$y'' + y = 0$$

Solutions are $y_1 = \cos x = Y_1, y_2 = \sin x = Y_2$

Another equation

$$y'' - y = 0$$

Solutions are $y_1 = e^x, y_2 = e^{-x}$

General solution is $c_1e^x + c_2e^{-x}$

We want to find $Y$

$Y_1: c_1 + c_2 = 1$

$c_1 - c_2 = 0$

$c_1 = c_2 = 1/2$

$Y_1 = \frac{e^x + e^{-x}}{2}$ which is called $\cosh x$

Same calculation, by using initial condition 0, 1

$Y_2 = \frac{e^x - e^{-x}}{2}$ which is called $\sinh x$

Why do we want normalized solutions?

If $Y_1, Y_2$ are normalized at 0, then solution to IVP ODE $y(0) = y_0, y'(0) = y_0'$ is $y = y_1Y_1 + y_1'Y_2$

$Y_1, Y_2$ enable us to write down the soluiton of IVP


## Answer of Q2

Existence and uniqueness theorem

$$y'' + py' + qy = 0$$

$p, q$ are continuous for all x

There is one and only one solution such that $y(0) = A, y'(0) = B$

Existence: The initial value problem has a solution

Uniqueness: It only has one solution

Claim: $c_1Y_1 + c_2Y_2$ are all solutions. Why?

Proof: Given a solution $u(x)$ where $u(0) = u_0, u'(0) = u'_0$, then $u_0Y_1 + u_0' Y_2$ satisfies these initial value conditions. The existence and uniqueness theorem said that there is only one solution so $u(x)$ must be equal to  $u_0Y_1 + u_0' Y_2$
