# Fourier Series

$$f(t) = \frac{a_0}{2} + \sum_{n=1}^\infty [a_n\cos nt + b_n \sin nt]$$

$$a_n = \frac{1}{\pi}\int_{-\pi}^{\pi} f(t) \cos nt dt$$

$$b_n = \frac{1}{\pi}\int_{-\pi}^{\pi} f(t) \sin nt dt$$

with $f(t)$ period $2\pi$

If $f(t) = g(t)$, then FS for f = FS for g

```julia
using Plots
plot(x->x, -pi, pi)
```

Suppose we have function above, which is not periodic. We can make it periodic with interval $[-\pi, \pi]$

1. Shorten calculation
2. Extend the range of Fourier Series

## Shorten calculation

Using even and odd function property

### Even function

$f(t)$ is even means $f(t) = f(-t)$. For example $\cos mt$ is even

If a function is even, its Fourier Series has only cosine terms

$$f(t) = \frac{a_n}{2} + \sum_{n=1}^\infty a_n\cos nt $$

because all of the $b_n = 0$. Why?

$$f(t) = \frac{a_0}{2} + \sum_{n=1}^\infty a_n\cos nt + b_n\sin nt$$

$$f(-t) = \frac{a_0}{2} + \sum_{n=1}^\infty a_n\cos nt - b_n\sin nt$$

$$f(t) = f(-t)$$

$$b_n = -b_n$$

Therefore $b_n$ must be zero

### Odd function

$f(t)$ is odd $\implies$ all $a_n = 0$. Same reasoning.



The fact that $f(t)$ is even can simplify the calculation of $a_n$

$f(t) \cos nt$ is also even. Integration of a function from $-\pi$ to $\pi$ is equal to 2 time integration from 0 to $\pi$

$f$ is Even:

$$a_n = \frac{2}{\pi}\int_0^{\pi} \cos nt dt, (b_n = 0)$$

$f$ is Odd, $\sin$ is Odd, Odd*Odd = Even function:

$$b_n = \frac{2}{\pi}\int_0^{\pi} \sin nt dt, (a_n = 0)$$

### Example

$f(t)$ is a sawtooth function. It is an odd function, so $a_n = 0$

$$b_n = \frac{2}{\pi}\int_0^{\pi} t\sin nt dt$$

Integrate by parts, copy first term multiply integral of second term. Then substract integral of differentiate first term multiply by integral of second term

$$b_n = \frac{2}{\pi}\int_0^{\pi} t\sin nt dt$$

$$= \frac{2}{\pi}[-t\frac{\cos nt}{n}|_0^{\pi} - \int_0^{\pi}\frac{-\cos nt}{n}dt]$$

$$= \frac{2}{\pi}[\frac{-\pi}{n}(-1)^n + \frac{\sin nt}{n^2}|_0^{\pi}]$$

$$b_n = \frac{2}{n}(-1)^{n+1}$$

Fourier Series for $$f(t) = \sum_1^\infty \frac{2(-1)^{n+1}}{n} \sin nt = 2(\sin t - \frac12 \sin 2t + \frac13 \sin 3t + ...)$$

First term approximate the function, then second, third term correct the error to make it better and better



```julia
using Plots
f(t) = 2*(sin(t) - 1/2*sin(2t) + 1/3*sin(3t) - 1/4*sin(4t) + 1/5*sin(5t))
plot(f, -pi, 4pi)  # approximate sawtooth function
```

Theorem: If $f$ is continuous at $t_0$, then $f(t)$ = sum of Fourier Series at $t_0$ (Fourier Series converges)

If $f$ has jumps discontinuity at $t_0$, then sum of Fourier Series converges to mid point of the jump

$$f(t) = \sum_1^\infty \frac{2(-1)^{n+1}}{n} \sin nt$$, 

at n = 0, $f(t) = 0$

## Extension 1 - Changing Period

Let say original interval for u is $[0, \pi]$, to convert t which has the interval $[0, L]$, 

$$t = \frac{L}{\pi} u$$

$$u = \frac{\pi}{L} t$$

$\cos nt \to \cos \frac{n\pi t}{L}$, $\sin nt \to \sin \frac{n\pi t}{L}$

zero at $0, \pm L, \pm 2L, ...$

$$f(t) = \frac{a_0}{2} + \sum_{n=1}^\infty [a_n\cos \frac{n\pi }{L}t + b_n \sin \frac{n\pi}{L}t]$$

$$a_n = \frac{1}{L}\int_{-L}^{L} f(t) \cos \frac{n\pi }{L}t dt$$

$$b_n = \frac{1}{L}\int_{-L}^{L} f(t) \sin \frac{n\pi }{L}t dt$$

### Period 2L

$f(t)$ is even, period $2L$

$$a_n = \frac{2}{L}\int_{0}^{L} f(t) \cos \frac{n\pi }{L}t dt, b_n = 0$$

etc for $f(t)$ is odd

## Extension 2

For function on finite interval, use Fourier Series. Fourier transform for infinite interval

Let say $f(t)$ defined on $[0, L]$. The way to apply FS is to make the periodic extension

```julia
L = 1
plot(x->x^2, 0, L) # this function only defined [0, 1]
```

```julia
# extended to be an even function
plot(x->x^2, -L, L)
```

```julia
# odd extension
f(x) = x^2
plot(f, 0, L)
```

```julia
plot!(x->-x^2, -L, 0)
```
