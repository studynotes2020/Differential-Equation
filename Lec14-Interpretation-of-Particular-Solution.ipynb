# Recall

$$y'' + ay' + by = f(t)$$

$$(D^2 + aD + b)y = f(t)$$

$$p(D)y = f(t)$$

where $p(D) = D^2 + aD + b$

$p(D)$ can be thought as formal polynomial in D. It is also a linear operator on $y(t)$

We are only interested in exponential, $f(t) = e^{at}$, a is complex

We want to find particular solution

$$y_p = \frac{e^{at}}{p(a)}$$

where $p(a) \neq 0$

If $p(a) = 0$, but $p'(a) \neq 0$ (a is simple root), then

$$y_p = \frac{te^{at}}{p'(a)}$$



# Resonance

$$y'' + w_0^2y = \cos w_1t$$

$w_1 \neq w_0$ driving the system with the input frequency not equal to the natural frequency

$$(D^2 + w_0^2)y = \cos w_1t$$

Complexify

$$(D^2 + w_0^2)\tilde y = e^{iw_1t}$$

$$\tilde y_p = \frac{e^{iw_1t}}{(iw_1)^2 + w_0^2} = \frac{e^{iw_1t}}{w_0^2 - w_2^2}$$

$$Re(\tilde y_p) = y_p = \frac{\cos w_1t}{w_0^2 - w_1^2}$$

$\tilde y_p$ is the response. Interestingly, it is almost the same as input, except $w_0^2 - w_1^2$ at the denominator which means the amplitude is changed

When $w_1 \approx w_0$, amplitude is bigger than then input with the same frequency

Suppose $w_1 = w_0$, then $p(a) \neq 0$ is not valid, so we should use the second version of exponential shift law

$$(D^2 + w_0^2)y = \cos w_0t$$

$iw_0$ is simple zero of $D^2 + w_0^2$, therefore

$$\tilde y_p = \frac{te^{iw_0t}}{2iw_0}$$

where $p'(a) = 2D$

$$Re(\tilde y_p) = y_p = \frac{t\sin w_0t}{2w_0}$$

How does the first solution turns into the second solution? (They look very different)

That is not the only particular solution. Any solution to the inhomogeneous differential equation is a particular solution. All the solutions are special, we don't need to use that one. Where does all other solutions?

$$(D^2 + w_0^2)y = \cos w_1t$$

$$y_p = \frac{\cos w_1t}{w_0^2 - w_1^2} - \frac{\cos w_0t}{w_0^2 - w_1^2}$$

Second term from complementary solution, steal it from there, make it into particular solution

Calculate the limit of the expression

$$\lim\limits_{w_1 \to w_0} \frac{\cos w_1t - \cos w_0t}{w_0^2 - w_1^2}$$

If we don't add the second term, $y_p \to \infty$, does not gives us anything. But after we add the second term, we can use Lhopital rule

$$= \lim\limits_{w_1 \to w_0} \frac{-t\sin(w_1t)}{-2w_1} = \frac{t\sin w_0t}{2w_0} $$


### Geometric meaning

Trigonometry identity

$\cos B - \cos A = 2\sin \frac{A - B}{2} \sin \frac{A + B}{2}$

$$\frac{\cos w_1t - \cos w_0t}{w_0^2 - w_1^2} = \frac{2}{w_0^2 - w_1^2} \sin (\frac{w_0 - w_1}{2}t) \sin (\frac{w_0 + w_1}{2}t)$$

$w_1 \approx w_0$

$$\frac{2}{w_0^2 - w_1^2} \sin (\frac{w_0 - w_1}{2}t) \sin (w_0t)$$

First sine, the frequency is small and amplitude is varying while second sine is purely oscillation with frequency relative large compare to the first one

This is called beat

## Damped resonance

$$mx'' + cx' + kx = f_1(t)$$

$$x'' + 2px' + w_0^2x = f_2(t)$$

$w_0$ is the natural undamped frequency. If $2p$ is not there, it is a system that the spring vibrates by itself

Visual: $$x'' + bx' + k_2x = f_3(t)$$

$w_1$ is natural damped frequency, "pseudofrequency". Function isn't periodic, it decays, nonetheless, it still cross the t-axis on regular interval

The relation of natural frequency and damped frequency is

$$p^2 + w_1^2 = w_0^2 \implies w_1 = \sqrt{w_0^2 - p^2}$$

$w_0^2$ is fixed by the string, so the bigger the damping $p$, the smaller $w_1$ is. Also, if there is little damping, the damped frequency is almost the same as natural frequency. This comes from the characteristic roots of the damped equation

If you got a damped spring, what happens when you impose a motion on it with yet thrid frequency

$$y'' + 2py' + w_0^2y = \cos wt$$

The problem is which $w$ gives maximal amplitude for response? WHen $w = w_0$. But when it is damped the answer is different. 

Given specific value of $p, w_0$. The answer is resonant frequency $w_r = \sqrt{w_0^2 - 2p^2}$
