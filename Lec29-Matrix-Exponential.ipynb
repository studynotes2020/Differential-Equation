# Fundamental Matrix

$$\vec x' = A\vec x$$

Fundamental matrix for A: $$X = \begin{bmatrix}\vec x_1 & \vec x_2 \end{bmatrix}$$

x1, x2 are two independent solutions

Basic properties of X

1. Determinant, $|X| \neq 0$ for any t
2. $X' = AX$ columns solve the system

Fundamental matrix A is not unique. There are infinity of ways to pick independent solutions

Write general solutions to system

$$\vec x = c_1\vec x_1 + c_2 \vec x_2$$

$$\vec x = X \vec c = \begin{bmatrix}\vec x_1 & \vec x_2 \end{bmatrix}
\begin{bmatrix} c_1 \\ c_2 \end{bmatrix} $$

What do all fundamental matrix look like?

$$\begin{bmatrix} X \vec c_1 & X \vec c_2 \end{bmatrix} =
X\begin{bmatrix} \vec c_1 & \vec c_2 \end{bmatrix} = XC $$

The most general fundamental matrix is $XC$

Both determinant of X and C is not zero

## Formula to get solutions

Lecture till now discuss about the method to find a solution, not a formula. Now we will be looking for the formula

1x1 case: For $x' = ax$, solution is $x = ce^{at}$. This is the formula

Definition of exponential comes from infinite series
$$e^{at} = 1 + at + \frac{a^2t^2}{2!} + \frac{a^3t^3}{3!} + ...$$

$$\frac{d(e^{at})}{dt} = a + a^2t + \frac{a^3t^2}{2!} + ... = ae^{at}$$

By differentiating term by term, we can see it satisfy $x' = ax$

How about 2x2 case?

A fundamental matrix for $\vec x' = A\vec x$  is 
$$e^{At} := I + At + \frac{A^2t^2}{2!} + \frac{A^3t^3}{3!} + ...$$

A is 2x2 matrix. The result of multiplying A by itself and with a scalar then sum them up is still a 2x2 matrix

Is this a fundamental matrix for the system?

Yes, $X = e^{At}$ satisfies $X' = AX$. Just check determinant at t = 0, which is determinant of identity, is non-zero


## Example

$$x' = y$$
$$y' = x$$

Therefore, $A = \begin{bmatrix}0 & 1 \\ 1 & 0 \end{bmatrix}$

$$\vec x' = \begin{bmatrix}0 & 1 \\ 1 & 0 \end{bmatrix} \vec x$$

$A^2 = \begin{bmatrix}0 & 1 \\ 1 & 0 \end{bmatrix} \begin{bmatrix}0 & 1 \\ 1 & 0 \end{bmatrix} = \begin{bmatrix}0 & 1 \\ 1 & 0 \end{bmatrix} = I$

$A^3 = A^2A = IA = A$

$A^4 = A^3A = AA = I$

$$e^{At} = I + \begin{bmatrix}0 & 1 \\ 1 & 0 \end{bmatrix}t + I\frac{t^2}{2!} + \begin{bmatrix}0 & 1 \\ 1 & 0 \end{bmatrix} \frac{t^3}{3!} + ...$$

$$e^{At} = \begin{bmatrix}1 + \frac{t^2}{2!} + \frac{t^4}{4!} + ... & t + \frac{t^3}{3!} + ... \\ t + \frac{t^3}{3!} + ... & 1 + \frac{t^2}{2!} + \frac{t^4}{4!} + ...\end{bmatrix}$$

$$e^{At} = \begin{bmatrix}\cosh t & \sinh t \\ \sinh t & \cosh t\end{bmatrix} = \frac12 \begin{bmatrix}e^t + e^{-t} & e^t - e^{-t} \\ e^t - e^{-t} & e^t + e^{-t} \end{bmatrix}$$

## Initial Value Problem

$$\vec x' = A\vec x$$

Initial value $\vec x(0) = \vec x_0$. Find $\vec x(t)$

General Solutions: $$\vec x = e^{At} \vec c$$

Initial value:  $$\vec x(0) = e^{A0}\vec c$$

$$\vec x_0 = I \vec c = \vec c$$

Solutions: $$\vec x = e^{At}\vec x_0$$

We cannot put $x_0$ in front because it is column vector (nx1) and it will multiply $e^{At}$ a nxn matrix, which gives invalid matrix multiplication

However matrix exponential doesn't satisfy exponential law

$$e^{A + B} \neq e^A e^B$$

True only if $AB = BA$. Matrix multiplication is not commutative. Only three special cases

1. $A = cI = \begin{bmatrix} c & 0 \\ 0 & c \end{bmatrix}$
2. $B = -A$
3. $B = A^{-1}$

Hence this is true,

$$e^{A-A} = e^A e^{-A} = I$$

and shows that $$(e^A)^{-1} = e^{-A}$$

### Calculate $e^{At}$

1. Use series but too hard
2. Useful only for symmetric matrices. Breaking it up.

$$\begin{bmatrix} a & b \\ b & a \end{bmatrix} = 
\begin{bmatrix} a & 0 \\ 0 & a \end{bmatrix} + 
\begin{bmatrix} 0 & b \\ b & 0 \end{bmatrix}$$

Notice that this is the first case above, so we can use exponential law

3. All other cases. Starts with arbitrary fundamental matrix, multiply it with its inverse at zero

$$X.X(0)^{-1} = e^{At}$$

After multiplication, it is still fundamental matrix. And value of X at zero is identity.
